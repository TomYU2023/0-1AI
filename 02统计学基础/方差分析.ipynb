### 方差分析种类

| 类型                                    | 用法                                 | 区别              | 优点                                 | 缺点                       |
|---------------------------------------|------------------------------------|-----------------|------------------------------------|--------------------------|
| 单因素方差分析（One-way ANOVA）         | 比较一个因素在多个水平上的均值差异              | 单个因素            | 简单直观，易于理解                      | 无法处理多个因素或协变量              |
| 双因素方差分析（Two-way ANOVA）         | 比较两个因素及其交互作用对因变量的影响           | 两个因素及交互作用       | 可以同时分析两个因素的影响                | 分析较复杂，需要较大样本量             |
| 重复测量方差分析（Repeated Measures ANOVA） | 比较同一受试者在不同条件下的均值差异             | 同一受试者多次测量       | 适用于重复测量数据，控制个体差异             | 需要重复测量数据，设计复杂              |
| 协方差分析（ANCOVA）                   | 在方差分析中控制协变量的影响                   | 控制协变量           | 提高分析精度，控制协变量影响               | 需要协变量的数据，分析复杂              |
| 多元方差分析（MANOVA）                  | 比较多个因变量在多个组别之间的均值差异            | 多个因变量           | 考虑多个因变量的相关性                   | 分析较复杂，需要较大样本量             |
| 嵌套方差分析（Nested ANOVA）            | 分析嵌套结构中的变异                       | 嵌套结构            | 适用于嵌套设计，分析复杂结构                | 设计和分析复杂                    |



在统计学和实验设计中，以下是一些关键术语的定义：

1. **因素（Factor）**：
   - 因素是指在实验或分析中研究的独立变量，通常是你感兴趣的变量，用来观察其对因变量（响应变量）的影响。
   - 示例：在研究不同教学方法对学生成绩的影响时，教学方法就是一个因素。

2. **水平（Level）**：
   - 水平是指因素的不同类别或状态，每个因素可以有多个水平。
   - 示例：如果教学方法是因素，那么不同的教学方法如传统讲授法、小组讨论法和线上学习法就是这个因素的不同水平。

3. **协变量（Covariate）**：
   - 协变量是指在分析中被控制或调整的变量，虽然它们不是主要的研究对象，但它们可能对因变量有影响。加入协变量的目的是消除这些变量的影响，从而更准确地评估主要因素的效应。
   - 示例：在研究教学方法对学生成绩的影响时，学生的初始成绩可以作为一个协变量，因为它可能会影响最终的成绩。

### 示例说明

假设你正在研究不同教学方法（因素）对学生考试成绩（因变量）的影响：
- **因素**：教学方法
  - **水平**：传统讲授法、小组讨论法、线上学习法
- **因变量**：学生考试成绩
- **协变量**：学生的初始成绩

在单因素方差分析中，你会比较不同教学方法（因素的各个水平）对学生成绩的影响，而在协方差分析中，你会同时考虑学生的初始成绩（协变量），以便更准确地评估教学方法对成绩的影响。

以下是各类方差分析的实例说明：

### 1. 单因素方差分析（One-way ANOVA）
**实例**：比较不同教学方法对学生考试成绩的影响
- **因素**：教学方法
- **水平**：传统讲授法、小组讨论法、线上学习法
- **因变量**：学生考试成绩

**问题**：三种不同的教学方法对学生成绩有显著差异吗？

### 2. 双因素方差分析（Two-way ANOVA）
**实例**：研究教学方法和学生性别对考试成绩的影响
- **因素1**：教学方法（传统讲授法、小组讨论法、线上学习法）
- **因素2**：性别（男、女）
- **因变量**：学生考试成绩

**问题**：教学方法和性别对学生成绩有显著影响吗？是否存在教学方法和性别之间的交互作用？

### 3. 重复测量方差分析（Repeated Measures ANOVA）
**实例**：测量同一组学生在不同时间点的考试成绩
- **因素**：时间点
- **水平**：期中考试成绩、期末考试成绩
- **因变量**：学生成绩

**问题**：学生在期中和期末的考试成绩有显著差异吗？

### 4. 协方差分析（ANCOVA）
**实例**：比较不同教学方法对学生成绩的影响，同时控制学生的初始成绩
- **因素**：教学方法
- **水平**：传统讲授法、小组讨论法、线上学习法
- **因变量**：学生期末考试成绩
- **协变量**：学生期初成绩

**问题**：不同教学方法在控制初始成绩后，对期末成绩有显著影响吗？

### 5. 多元方差分析（MANOVA）
**实例**：比较不同教学方法对学生成绩和学习满意度的综合影响
- **因素**：教学方法
- **水平**：传统讲授法、小组讨论法、线上学习法
- **因变量**：学生成绩、学习满意度

**问题**：不同教学方法对学生成绩和满意度的综合影响有显著差异吗？

### 6. 嵌套方差分析（Nested ANOVA）
**实例**：研究不同学校和学校内部不同班级对学生成绩的影响
- **因素1**：学校
- **因素2**：班级（嵌套在学校内）
- **因变量**：学生成绩

**问题**：不同学校之间和学校内部不同班级之间的学生成绩有显著差异吗？

这些实例说明了每种方差分析方法在不同情境中的具体应用。通过合理选择和应用这些方法，可以更准确地分析实验数据，得出有意义的结论。

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.multivariate.manova import MANOVA

# Generate sample data
np.random.seed(0)

# Example data
data = pd.DataFrame({
    'method': np.repeat(['traditional', 'group_discussion', 'online'], 30),
    'gender': np.tile(np.repeat(['male', 'female'], 15), 3),
    'time': np.tile(np.repeat(['midterm', 'final'], 15), 3),
    'initial_score': np.random.normal(70, 10, 90),
    'score': np.concatenate([
        np.random.normal(70, 10, 30),
        np.random.normal(75, 10, 30),
        np.random.normal(80, 10, 30)
    ])
})

# Add different time points measurements
data['score'] += np.tile(np.random.normal(5, 3, 30), 3)

# One-way ANOVA
model1 = ols('score ~ C(method)', data=data).fit()
anova_table1 = sm.stats.anova_lm(model1, typ=2)

# Two-way ANOVA
model2 = ols('score ~ C(method) * C(gender)', data=data).fit()
anova_table2 = sm.stats.anova_lm(model2, typ=2)

# Repeated Measures ANOVA
model3 = ols('score ~ C(time)', data=data).fit()
anova_table3 = sm.stats.anova_lm(model3, typ=2)

# ANCOVA
model4 = ols('score ~ C(method) + initial_score', data=data).fit()
anova_table4 = sm.stats.anova_lm(model4, typ=2)

# MANOVA
data['satisfaction'] = np.random.normal(7, 1.5, 90)
manova = MANOVA.from_formula('score + satisfaction ~ C(method)', data=data)
manova_results = manova.mv_test()

# Nested ANOVA
data['school'] = np.repeat(['A', 'B', 'C'], 30)
data['class_group'] = np.tile(np.repeat(['class1', 'class2', 'class3'], 10), 3)
model6 = ols('score ~ C(school) / C(class_group)', data=data).fit()
anova_table6 = sm.stats.anova_lm(model6, typ=2)

# Results and plots output
def plot_anova_results(anova_table, title):
    print(f"\n{title}\n", anova_table)
    anova_table.plot(kind='bar', figsize=(10, 6), legend=False)
    plt.title(title)
    plt.show()

# Output results and plots
plt.figure(figsize=(10, 6))
sns.boxplot(x='method', y='score', data=data)
plt.title('One-way ANOVA: Method on Score')
plt.show()

print("\nOne-way ANOVA\n", anova_table1)
```


### 1. 单因素方差分析（One-way ANOVA）
- **问题**：不同教学方法（traditional, group_discussion, online）对学生成绩有显著影响吗？
- **结果**：教学方法对学生成绩的影响不显著（p > 0.05）。


```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='method', y='score', hue='gender', data=data)
plt.title('Two-way ANOVA: Method and Gender on Score')
plt.show()

print("\nTwo-way ANOVA\n", anova_table2)
```



### 2. 双因素方差分析（Two-way ANOVA）
- **问题**：教学方法和性别对学生成绩有显著影响吗？是否存在交互作用？
- **结果**：教学方法和性别对学生成绩的单独影响不显著（p > 0.05），也没有显著的交互作用（p > 0.05）。


```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='time', y='score', data=data)
plt.title('Repeated Measures ANOVA: Time on Score')
plt.show()

print("\nRepeated Measures ANOVA\n", anova_table3)
```



### 3. 重复测量方差分析（Repeated Measures ANOVA）
- **问题**：在不同时间点（midterm, final）学生成绩有显著差异吗？
- **结果**：时间对学生成绩的影响接近显著（p = 0.056）。


```python

plt.figure(figsize=(10, 6))
sns.lmplot(x='initial_score', y='score', hue='method', data=data, aspect=2)
plt.title('ANCOVA: Method on Score (controlling for Initial Score)')
plt.show()

print("\nANCOVA\n", anova_table4)
```



### 4. 协方差分析（ANCOVA）
- **问题**：在控制初始成绩后，不同教学方法对学生成绩有显著影响吗？
- **结果**：教学方法对学生成绩的影响不显著（p > 0.05）。



```python

plt.figure(figsize=(10, 6))
sns.pointplot(x='method', y='score', data=data, capsize=.2)
sns.pointplot(x='method', y='satisfaction', data=data, capsize=.2, color='orange')
plt.title('MANOVA: Method on Score and Satisfaction')
plt.show()

print("\nMANOVA\n", manova_results)
```


### 5. 多元方差分析（MANOVA）
- **问题**：不同教学方法对学生成绩和满意度的综合影响有显著差异吗？
- **结果**：教学方法对学生成绩和满意度的综合影响不显著（p > 0.05）。


```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='school', y='score', hue='class_group', data=data)
plt.title('Nested ANOVA: School and Class on Score')
plt.show()

print("\nNested ANOVA\n", anova_table6)
```



### 6. 嵌套方差分析（Nested ANOVA）
- **问题**：不同学校和学校内部不同班级对学生成绩有显著影响吗？
- **结果**：学校和班级对学生成绩的影响不显著（p > 0.05）。
