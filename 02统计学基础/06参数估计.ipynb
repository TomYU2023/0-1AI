- ÈÄâÊã©Ê≠£ÊÄÅÂàÜÂ∏ÉÊ®°ÂûãÂÅáËÆæÂ≠¶ÁîüÊàêÁª©Êúç‰ªéÊ≠£ÊÄÅÂàÜÂ∏É„ÄÇ(ÂΩìÁÑ∂‰πüÂèØ‰ª•ÂÅáËÆæÂÖ∂‰ªñÊ®°ÂûãÂÅáËÆæÔºåÂ¶Ç‰∫åÈ°πÂàÜÂ∏É„ÄÅÁ¶ªÊï£ÂàÜÂ∏ÉÁ≠â)



#### Ê≠•È™§2ÔºöÊî∂ÈõÜÊ†∑Êú¨Êï∞ÊçÆ
- ‰ªéÁè≠Á∫ß‰∏≠ÈöèÊú∫ÊäΩÂèñ‰∏Ä‰∏™Ê†∑Êú¨ÔºåÂÅáËÆæÊàë‰ª¨ÊäΩÂèñ‰∫Ü10ÂêçÂ≠¶ÁîüÁöÑÊàêÁª©Ôºö
  $X = [85, 90, 78, 92, 88, 76, 84, 95, 80, 82]$



#### Ê≠•È™§3ÔºöËÆ°ÁÆóÊ†∑Êú¨ÁªüËÆ°Èáè
- Ê†∑Êú¨ÂùáÂÄº$\bar{X}$Ôºö
  $$
  \bar{X} = \frac{\sum_{i=1}^{n} X_i}{n} = \frac{85 + 90 + 78 + 92 + 88 + 76 + 84 + 95 + 80 + 82}{10} = 85
  $$
- Ê†∑Êú¨Ê†áÂáÜÂ∑Æ$s$Ôºö
  $$
  s = \sqrt{\frac{\sum_{i=1}^{n} (X_i - \bar{X})^2}{n-1}} = \sqrt{\frac{(85-85)^2 + (90-85)^2 + \cdots + (82-85)^2}{9}} \approx 6.27
  $$



#### Ê≠•È™§4ÔºöÈÄâÊã©‰º∞ËÆ°ÊñπÊ≥ï
- ËÆ°ÁÆóÁÇπ‰º∞ËÆ°ÂÄº„ÄÇ
- ÈÄâÊã©Âå∫Èó¥‰º∞ËÆ°ÁöÑÊñπÊ≥ï„ÄÇ


##### Âå∫Èó¥‰º∞ËÆ°

| Ê±ÇŒºËøòÊòØœÉ | œÉ   | ÁΩÆ‰ø°Ê∞¥Âπ≥ | Ê±Ç‰ªÄ‰πà       | Á≠îÊ°à                                                                                      | 
|----------|-----|----------|--------------|-------------------------------------------------------------------------------------------
| Ê±ÇŒº      | Â∑≤Áü• | 1-Œ±      | ÁΩÆ‰ø°Âå∫Èó¥     | $$ \left( \bar{X} - \frac{\sigma}{\sqrt{n}}z_{\frac{\alpha}{2}}, \bar{X} + \frac{\sigma}{\sqrt{n}}z_{\frac{\alpha}{2}} \right) $$ |

|          |     |          | Âçï‰æßÁΩÆ‰ø°‰∏äÈôê | $$ \bar{X} + \frac{\sigma}{\sqrt{n}}z_{\alpha} $$                                                                                              |
| Ê±ÇŒº      | Êú™Áü• | 1-Œ±      | ÁΩÆ‰ø°Âå∫Èó¥     | $$ \left( \bar{X} - \frac{S}{\sqrt{n}}t_{\frac{\alpha}{2}}(n-1), \bar{X} + \frac{S}{\sqrt{n}}t_{\frac{\alpha}{2}}(n-1) \right) $$ |                                                    
|          |     |          | Âçï‰æßÁΩÆ‰ø°‰∏ãÈôê | $$ \bar{X} - \frac{S}{\sqrt{n}}t_{\alpha}(n-1) $$                                                                                              |
|          |     |          | Âçï‰æßÁΩÆ‰ø°‰∏äÈôê | $$ \bar{X} + \frac{S}{\sqrt{n}}t_{\alpha}(n-1) $$                                          |                                                    
| Ê±ÇœÉ      | Êú™Áü• | 1-Œ±      | ÁΩÆ‰ø°Âå∫Èó¥     | $$ \left( \sqrt{\frac{(n-1) \times S^2}{\chi^2_{\frac{\alpha}{2}}(n-1)}}, \sqrt{\frac{(n-1) \times S^2}{\chi^2_{1-\frac{\alpha}{2}}(n-1)}} \right) $$ |                                                    
|          |     |          | Âçï‰æßÁΩÆ‰ø°‰∏ãÈôê | $$ \sqrt{\frac{(n-1) \times S^2}{\chi^2_{\alpha}(n-1)}} $$                                 |                                                    
|          |     |          | Âçï‰æßÁΩÆ‰ø°‰∏äÈôê | $$ \sqrt{\frac{(n-1) \times S^2}{\chi^2_{1-\alpha}(n-1)}} $$                               |                                                    

**Â§áÊ≥®**

$ z_{\alpha} „ÄÅ t_{\alpha} Âíå \chi^2_{\alpha}$ ÂùáÂèØÈÄöËøáÊü•Ë°®Ëé∑Âæó„ÄÇ

$ S^2 = \frac{(x_1 - \bar{X})^2 + (x_2 - \bar{X})^2 + \cdots + (x_n - \bar{X})^2}{n-1} $




#### Ê≠•È™§5ÔºöËÆ°ÁÆóÁÇπ‰º∞ËÆ°ÂÄº
- ‰ΩøÁî®Ê†∑Êú¨ÂùáÂÄº$\bar{X}$‰Ωú‰∏∫ÊÄª‰ΩìÂùáÂÄº$\mu$ÁöÑÁÇπ‰º∞ËÆ°ÂÄºÔºö
  $$
  \hat{\mu} = \bar{X} = 85
  $$



#### Ê≠•È™§6ÔºöÊûÑÂª∫ÁΩÆ‰ø°Âå∫Èó¥ÔºàÂå∫Èó¥‰º∞ËÆ°Ôºâ
- ÈÄâÊã©95%ÁöÑÁΩÆ‰ø°Ê∞¥Âπ≥ÔºåÊü•ÊâæÂØπÂ∫îÁöÑtÂàÜÂ∏É‰∏¥ÁïåÂÄºÔºàËá™Áî±Â∫¶‰∏∫$n-1 = 9$Ôºâ„ÄÇ
- ‰æãÂ¶ÇÔºåÂØπ‰∫é95%ÁöÑÁΩÆ‰ø°Ê∞¥Âπ≥ÔºåËá™Áî±Â∫¶‰∏∫9ÔºåtÂàÜÂ∏É‰∏¥ÁïåÂÄº$t_{0.025} \approx 2.262$„ÄÇ
- ÊûÑÂª∫ÁΩÆ‰ø°Âå∫Èó¥Ôºö
  $$
  \left( \bar{X} - t_{0.025} \frac{s}{\sqrt{n}}, \bar{X} + t_{0.025} \frac{s}{\sqrt{n}} \right) = \left( 85 - 2.262 \cdot \frac{6.27}{\sqrt{10}}, 85 + 2.262 \cdot \frac{6.27}{\sqrt{10}} \right)
  $$
- ËÆ°ÁÆóÁΩÆ‰ø°Âå∫Èó¥‰∏ä‰∏ãÈôêÔºö
  $$
  \left( 85 - 4.48, 85 + 4.48 \right) = \left( 80.52, 89.48 \right)
  $$


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given data
mean = 85
s = 6.27
n = 10
df = n - 1
confidence_level = 0.95
t_critical = 2.262  # from the t-distribution table for df = 9 and 95% confidence level

# Calculate the margin of error
margin_of_error = t_critical * (s / np.sqrt(n))

# Calculate the confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
confidence_interval = (round(confidence_interval[0], 2), round(confidence_interval[1], 2))

# Plot the sample data distribution and confidence interval
x = np.linspace(mean - 4 * s, mean + 4 * s, 1000)
y = stats.t.pdf(x, df, mean, s / np.sqrt(n))

plt.figure(figsize=(10, 6))

# Plot the t-distribution
plt.plot(x, y, label='t-distribution (df=9)', color='blue')

# Highlight the confidence interval
plt.fill_between(x, 0, y, where=(x >= confidence_interval[0]) & (x <= confidence_interval[1]), 
                 color='gray', alpha=0.5, label=f'95% Confidence Interval {confidence_interval}')

# Mark the mean
plt.axvline(mean, color='red', linestyle='--', label='Sample Mean')

# Mark the confidence interval limits
plt.axvline(confidence_interval[0], color='green', linestyle='--', label=f'Lower Limit ({confidence_interval[0]})')
plt.axvline(confidence_interval[1], color='green', linestyle='--', label=f'Upper Limit ({confidence_interval[1]})')

# Add labels and legend
plt.title('95% Confidence Interval for the Sample Mean')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

plt.show()
```

### 1. Ëá™ÂèòÈáèÂíåÂõ†ÂèòÈáèÈÉΩÊòØËøûÁª≠ÂèòÈáèÔºöÂõûÂΩíÂàÜÊûê

**ÁªüËÆ°Èáè**Ôºö
- **Ê†∑Êú¨Âπ≥ÂùáÊï∞**„ÄÅ**Ê†∑Êú¨ÊñπÂ∑Æ**ÔºöÊèèËø∞Ëá™ÂèòÈáèÂíåÂõ†ÂèòÈáèÁöÑÊï∞ÊçÆÂàÜÂ∏ÉÊÉÖÂÜµ„ÄÇ
- **ÂõûÂΩíÁ≥ªÊï∞ÔºàRegression CoefficientsÔºâ**ÔºöÂõûÂΩíÊñπÁ®ãÁöÑÂèÇÊï∞ÔºåÂ¶ÇÊñúÁéáÂíåÊà™Ë∑ù„ÄÇ

**‰º∞ËÆ°Èáè**Ôºö
- **ÂõûÂΩíÁ≥ªÊï∞ÁöÑ‰º∞ËÆ°Èáè**ÔºöÂ¶ÇÊúÄÂ∞è‰∫å‰πòÊ≥ïÔºàOLSÔºâ‰º∞ËÆ°ÂõûÂΩíÁ≥ªÊï∞„ÄÇ

**ÁÇπ‰º∞ËÆ°**Ôºö
- **ÂõûÂΩíÁ≥ªÊï∞ÁöÑÁÇπ‰º∞ËÆ°**ÔºöÈÄöËøáÊ†∑Êú¨Êï∞ÊçÆËÆ°ÁÆóÂæóÂà∞ÁöÑÂõûÂΩíÁ≥ªÊï∞ÂÖ∑‰ΩìÂÄºÔºàÂ¶ÇÊñúÁéáÂíåÊà™Ë∑ùÔºâ„ÄÇ

**Âå∫Èó¥‰º∞ËÆ°**Ôºö
- **ÂõûÂΩíÁ≥ªÊï∞ÁöÑÁΩÆ‰ø°Âå∫Èó¥**ÔºöÁªôÂá∫ÂõûÂΩíÁ≥ªÊï∞ÁöÑ‰∏Ä‰∏™ÂèØ‰ø°ËåÉÂõ¥„ÄÇ‰æãÂ¶ÇÔºåÈÄöËøáÊ†∑Êú¨Êï∞ÊçÆËÆ°ÁÆóÊñúÁéáÁöÑ95%ÁΩÆ‰ø°Âå∫Èó¥„ÄÇ

**‰æãÂ≠ê**Ôºö
ÂÅáËÆæÊàë‰ª¨Êúâ‰∏ÄÁªÑÊï∞ÊçÆÔºåË°®Á§∫ÂπøÂëäË¥πÁî®ÂíåÈîÄÂîÆÈ¢ù‰πãÈó¥ÁöÑÂÖ≥Á≥ªÔºåÊàë‰ª¨Â∞Ü‰ΩøÁî®Ëøô‰∫õÊï∞ÊçÆËøõË°åÂõûÂΩíÂàÜÊûêÔºåÂπ∂Â±ïÁ§∫ÁΩÆ‰ø°Âå∫Èó¥„ÄÇ


```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# ËÆæÁΩÆÈöèÊú∫ÁßçÂ≠ê‰ª•‰øùËØÅÁªìÊûúÂèØÈáçÂ§ç
np.random.seed(0)

# Á§∫‰æãÊï∞ÊçÆ
# ÂπøÂëäË¥πÁî®ÔºàÂçÉÂÖÉÔºâ
advertising_expenses = np.linspace(10, 55, 20)
# ÂØπÂ∫îÁöÑÈîÄÂîÆÈ¢ùÔºà‰∏áÂÖÉÔºâÔºåÊ∑ªÂä†‰∏Ä‰∫õÈöèÊú∫Âô™Â£∞
sales = 3 * advertising_expenses + 5 + np.random.normal(0, 10, advertising_expenses.shape)

# Ê∑ªÂä†Â∏∏Êï∞È°π
X = sm.add_constant(advertising_expenses)
model = sm.OLS(sales, X).fit()
predictions = model.predict(X)

# ËæìÂá∫ÂõûÂΩíÁªìÊûú
summary = model.summary()
print(summary)

# ÂèØËßÜÂåñ
plt.scatter(advertising_expenses, sales, label='Data points')
plt.plot(advertising_expenses, predictions, color='red', label='Regression line')

# ËÆ°ÁÆóÁΩÆ‰ø°Âå∫Èó¥
pred_std = np.sqrt(model.mse_resid)
confidence_interval = 1.96 * pred_std * np.sqrt(1/len(advertising_expenses) + (advertising_expenses - np.mean(advertising_expenses))**2 / np.sum((advertising_expenses - np.mean(advertising_expenses))**2))
predictions_lower = predictions - confidence_interval
predictions_upper = predictions + confidence_interval

# ÁªòÂà∂ÁΩÆ‰ø°Âå∫Èó¥
plt.fill_between(advertising_expenses, predictions_lower, predictions_upper, color='gray', alpha=0.2, label='95% Confidence interval')

plt.xlabel('Advertising Expenses (k$)')
plt.ylabel('Sales (10k$)')
plt.title('Advertising Expenses vs Sales Regression')
plt.legend()
plt.show()
```

### 2. Ëá™ÂèòÈáèÊòØÁ±ªÂà´ÂèòÈáèÔºåÂõ†ÂèòÈáèÊòØËøûÁª≠ÂèòÈáèÔºöÊñπÂ∑ÆÂàÜÊûêÔºàANOVAÔºâ

**ÁªüËÆ°Èáè**Ôºö
- **ÁªÑÈó¥ÊñπÂ∑Æ**Âíå**ÁªÑÂÜÖÊñπÂ∑Æ**ÔºöÁî®‰∫éÊèèËø∞‰∏çÂêåÁªÑÂà´ÁöÑÁ¶ªÊï£Á®ãÂ∫¶„ÄÇ
- **FÁªüËÆ°Èáè**ÔºöÁî®‰∫éÊ£ÄÈ™åÁªÑÈó¥ÂùáÂÄºÂ∑ÆÂºÇÁöÑÊòæËëóÊÄß„ÄÇ

**‰º∞ËÆ°Èáè**Ôºö
- **ÊñπÂ∑ÆÊØîÁöÑ‰º∞ËÆ°Èáè**ÔºöÈÄöËøáÊ†∑Êú¨Êï∞ÊçÆËÆ°ÁÆóÂæóÂà∞ÁªÑÈó¥ÊñπÂ∑ÆÂíåÁªÑÂÜÖÊñπÂ∑ÆÁöÑÊØîÂÄº„ÄÇ

**ÁÇπ‰º∞ËÆ°**Ôºö
- **ÂêÑÁªÑÂùáÂÄºÁöÑÁÇπ‰º∞ËÆ°**ÔºöÈÄöËøáÊ†∑Êú¨Êï∞ÊçÆËÆ°ÁÆóÂêÑÁªÑÁöÑÂπ≥ÂùáÂÄº„ÄÇ

**Âå∫Èó¥‰º∞ËÆ°**Ôºö
- **ÂêÑÁªÑÂùáÂÄºÂ∑ÆÂºÇÁöÑÁΩÆ‰ø°Âå∫Èó¥**ÔºöÁªôÂá∫ÁªÑÈó¥ÂùáÂÄºÂ∑ÆÂºÇÁöÑ‰∏Ä‰∏™ÂèØ‰ø°ËåÉÂõ¥„ÄÇ

**‰æãÂ≠ê**Ôºö
Á†îÁ©∂‰∏çÂêåÈ•ÆÈ£ü‰π†ÊÉØÂØπË°ÄÂéãÁöÑÂΩ±ÂìçÔºåÈÄöËøáÊñπÂ∑ÆÂàÜÊûêÂæóÂà∞ÂêÑÁªÑÂùáÂÄºÂ∑ÆÂºÇÁöÑFÁªüËÆ°ÈáèÔºåÂπ∂ËÆ°ÁÆóÂùáÂÄºÂ∑ÆÂºÇÁöÑÁΩÆ‰ø°Âå∫Èó¥„ÄÇ


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt

# Á§∫‰æãÊï∞ÊçÆ
data = {
    'Diet': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'BloodPressure': [120, 130, 115, 125, 135, 140, 130, 145, 150]
}

df = pd.DataFrame(data)

# ANOVAÂàÜÊûê
model = ols('BloodPressure ~ C(Diet)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# ÊâìÂç∞ÁªìÊûú
print(anova_table)

# Ëé∑ÂèñÂêÑÁªÑÂùáÂÄºÁöÑÁΩÆ‰ø°Âå∫Èó¥
means = df.groupby('Diet').mean()
se = df.groupby('Diet').std() / np.sqrt(df.groupby('Diet').count())
conf_int = sm.stats.multicomp.pairwise_tukeyhsd(endog=df['BloodPressure'], groups=df['Diet'], alpha=0.05)

# ÊâìÂç∞ÂêÑÁªÑÂùáÂÄºÁöÑÁΩÆ‰ø°Âå∫Èó¥
print(conf_int.summary())

# ÂèØËßÜÂåñ
df.boxplot('BloodPressure', by='Diet', grid=False)
for i, diet in enumerate(means.index):
    plt.errorbar(i+1, means.loc[diet, 'BloodPressure'], yerr=se.loc[diet, 'BloodPressure']*1.96, fmt='o', color='red', label='95% Confidence interval' if i == 0 else "")
plt.title('Diet vs BloodPressure')
plt.suptitle('')
plt.xlabel('Diet')
plt.ylabel('BloodPressure')
plt.legend()
plt.show()
```

### 3. Ëá™ÂèòÈáèÊòØËøûÁª≠ÂèòÈáèÔºåÂõ†ÂèòÈáèÊòØÁ±ªÂà´ÂèòÈáèÔºöÈÄªËæëÂõûÂΩí

- **ÈÄªËæëÂõûÂΩíÁ≥ªÊï∞**ÔºöÊèèËø∞Ëá™ÂèòÈáèÂØπÂõ†ÂèòÈáèÁöÑÂΩ±ÂìçÂº∫Â∫¶„ÄÇ

**‰º∞ËÆ°Èáè**Ôºö
- **ÈÄªËæëÂõûÂΩíÁ≥ªÊï∞ÁöÑ‰º∞ËÆ°Èáè**ÔºöÈÄöËøáÊúÄÂ§ß‰ººÁÑ∂‰º∞ËÆ°ÔºàMLEÔºâÊñπÊ≥ïËÆ°ÁÆóÂæóÂà∞ÈÄªËæëÂõûÂΩíÁ≥ªÊï∞„ÄÇ

**ÁÇπ‰º∞ËÆ°**Ôºö
- **ÈÄªËæëÂõûÂΩíÁ≥ªÊï∞ÁöÑÁÇπ‰º∞ËÆ°**ÔºöÈÄöËøáÊ†∑Êú¨Êï∞ÊçÆËÆ°ÁÆóÂæóÂà∞ÂÖ∑‰ΩìÁöÑÂõûÂΩíÁ≥ªÊï∞ÂÄº„ÄÇ

**Âå∫Èó¥‰º∞ËÆ°**Ôºö
- **ÈÄªËæëÂõûÂΩíÁ≥ªÊï∞ÁöÑÁΩÆ‰ø°Âå∫Èó¥**ÔºöÁªôÂá∫ÂõûÂΩíÁ≥ªÊï∞ÁöÑ‰∏Ä‰∏™ÂèØ‰ø°ËåÉÂõ¥„ÄÇ

**‰æãÂ≠ê**Ôºö
Á†îÁ©∂‰ΩìÈáçÊåáÊï∞ÔºàBMIÔºâÂØπÊòØÂê¶ÊÇ£ÊúâÈ´òË°ÄÂéãÁöÑÂΩ±ÂìçÔºåÈÄöËøáÈÄªËæëÂõûÂΩíÂàÜÊûêÂæóÂà∞ÈÄªËæëÂõûÂΩíÁ≥ªÊï∞ÁöÑÁÇπ‰º∞ËÆ°ÔºàÂ¶ÇBMIÁ≥ªÊï∞‰∏∫0.3ÔºâÔºåÂπ∂ËÆ°ÁÆóÂõûÂΩíÁ≥ªÊï∞ÁöÑÁΩÆ‰ø°Âå∫Èó¥ÔºàÂ¶Ç0.1Âà∞0.5Ôºâ„ÄÇ


```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# Âä†ËΩΩPima Indians DiabetesÊï∞ÊçÆÈõÜ
# Êï∞ÊçÆÈõÜÂåÖÂê´Ôºö'preg', 'plas', 'pres', 'skin', 'test', 'BMI', 'pedi', 'age', 'class'
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
data = pd.read_csv(url, names=columns)

# ÈÄâÊã©BMIÂíåOutcomeÂèòÈáè
df = data[['BMI', 'Outcome']]

# ÈÄªËæëÂõûÂΩí
X = sm.add_constant(df['BMI'])
y = df['Outcome']
model = sm.Logit(y, X).fit()

# ÊâìÂç∞ÁªìÊûú
print(model.summary())

# Ëé∑ÂèñÈÄªËæëÂõûÂΩíÁ≥ªÊï∞ÁöÑÁΩÆ‰ø°Âå∫Èó¥
conf = model.conf_int()
conf['Odds Ratio'] = model.params
conf.columns = ['5%', '95%', 'Odds Ratio']
print(conf)

# ÂèØËßÜÂåñ
plt.figure(figsize=(8, 6))

# Êï∞ÊçÆÁÇπÂíåÊãüÂêàÁ∫ø
x_vals = np.linspace(df['BMI'].min(), df['BMI'].max(), 300)
y_vals = model.predict(sm.add_constant(x_vals))
plt.plot(x_vals, y_vals, color='red', label='Logistic regression line')

# Ê∑ªÂä†Êï∞ÊçÆÁÇπ
plt.scatter(df['BMI'], df['Outcome'], alpha=0.5, label='Data points')

# ËÆ°ÁÆóÁΩÆ‰ø°Âå∫Èó¥
pred_std = np.sqrt(model.scale)
confidence_interval = 1.96 * pred_std * np.sqrt(1/len(df['BMI']) + (x_vals - np.mean(df['BMI']))**2 / np.sum((df['BMI'] - np.mean(df['BMI']))**2))
y_vals_lower = y_vals - confidence_interval
y_vals_upper = y_vals + confidence_interval

plt.fill_between(x_vals, y_vals_lower, y_vals_upper, color='gray', alpha=0.2, label='95% Confidence interval')

plt.xlabel('BMI')
plt.ylabel('Probability of Diabetes')
plt.title('BMI vs Diabetes Logistic Regression')
plt.legend()
plt.show()
```

### 4. Ëá™ÂèòÈáèÂíåÂõ†ÂèòÈáèÈÉΩÊòØÁ±ªÂà´ÂèòÈáèÔºöÂç°ÊñπÊ£ÄÈ™å

**ÁªüËÆ°Èáè**Ôºö
- **Âç°ÊñπÁªüËÆ°Èáè**ÔºöÁî®‰∫éÊ£ÄÈ™åÁ±ªÂà´ÂèòÈáèÈó¥ÁöÑÂÖ≥ËÅîÊÄß„ÄÇ

**‰º∞ËÆ°Èáè**Ôºö
- **ÊúüÊúõÈ¢ëÊï∞**Âíå**ËßÇÂØüÈ¢ëÊï∞**ÁöÑÊØîÁéá‰º∞ËÆ°Èáè„ÄÇ

**ÁÇπ‰º∞ËÆ°**Ôºö
- **ËßÇÂØüÈ¢ëÊï∞ÁöÑÁÇπ‰º∞ËÆ°**ÔºöÈÄöËøáÊ†∑Êú¨Êï∞ÊçÆËÆ°ÁÆóÂæóÂà∞ÂêÑÁ±ªÂà´ÁªÑÂêàÁöÑËßÇÂØüÈ¢ëÊï∞„ÄÇ

**Âå∫Èó¥‰º∞ËÆ°**Ôºö
- **Á±ªÂà´ÊØî‰æãÁöÑÁΩÆ‰ø°Âå∫Èó¥**ÔºöÁªôÂá∫ÂêÑÁ±ªÂà´ÊØî‰æãÁöÑ‰∏Ä‰∏™ÂèØ‰ø°ËåÉÂõ¥„ÄÇ

**‰æãÂ≠ê**Ôºö
Á†îÁ©∂ÊÄßÂà´ÂØπÊòØÂê¶ÂñúÊ¨¢ÊüêÁßçÈ•ÆÊñôÁöÑÂΩ±ÂìçÔºåÈÄöËøáÂç°ÊñπÊ£ÄÈ™åÂæóÂà∞Âç°ÊñπÁªüËÆ°ÈáèÔºåÂπ∂ËÆ°ÁÆóÂêÑÁ±ªÂà´ÊØî‰æãÁöÑÁΩÆ‰ø°Âå∫Èó¥„ÄÇ


```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency, norm
import matplotlib.pyplot as plt

# Á§∫‰æãÊï∞ÊçÆ
data = {
    'Gender': ['Male', 'Male', 'Female', 'Female'],
    'LikesDrink': ['Yes', 'No', 'Yes', 'No'],
    'Count': [20, 10, 30, 40]
}

df = pd.DataFrame(data)

# ÂàõÂª∫ÂàóËÅîË°®
contingency_table = pd.pivot_table(df, values='Count', index='Gender', columns='LikesDrink')

# Âç°ÊñπÊ£ÄÈ™å
chi2, p, dof, expected = chi2_contingency(contingency_table)

# ÊâìÂç∞ÁªìÊûú
print(f'Âç°ÊñπÁªüËÆ°Èáè: {chi2:.2f}')
print(f'pÂÄº: {p:.2f}')
print(f'Ëá™Áî±Â∫¶: {dof}')
print('ÊúüÊúõÈ¢ëÊï∞:\n', np.round(expected, 2))

# ËÆ°ÁÆóÂêÑÁ±ªÂà´ÊØî‰æãÁöÑÁΩÆ‰ø°Âå∫Èó¥
df['Proportion'] = df['Count'] / df['Count'].sum()
df['CI Lower'] = df['Proportion'] - 1.96 * np.sqrt((df['Proportion'] * (1 - df['Proportion'])) / df['Count'].sum())
df['CI Upper'] = df['Proportion'] + 1.96 * np.sqrt((df['Proportion'] * (1 - df['Proportion'])) / df['Count'].sum())

# ‰øùÁïôÂ∞èÊï∞ÁÇπÂêé‰∏§‰Ωç
df['Proportion'] = df['Proportion'].round(2)
df['CI Lower'] = df['CI Lower'].round(2)
df['CI Upper'] = df['CI Upper'].round(2)

# ÊâìÂç∞ÁªìÊûú
print(df)

# ÂèØËßÜÂåñ
labels = df['Gender'] + '-' + df['LikesDrink']
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(x, df['Proportion'], width, yerr=[df['Proportion'] - df['CI Lower'], df['CI Upper'] - df['Proportion']], capsize=5, label='Proportion with 95% CI')

ax.set_xlabel('Category')
ax.set_ylabel('Proportion')
ax.set_title('Proportion with 95% Confidence Intervals by Category')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
