```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import scipy.stats as stats
import matplotlib.pyplot as plt

# 
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

# 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Logistic
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

# 
y_pred_prob = model.predict_proba(X_test)[:, 1]

# Hosmer-Lemeshow 
def hosmer_lemeshow_test(y_true, y_pred, g=10):
    data = pd.DataFrame({'true': y_true, 'pred': y_pred})
    data['group'] = pd.qcut(data['pred'], g, duplicates='drop')
    observed = data.groupby('group', observed=False)['true'].sum()
    expected = data.groupby('group', observed=False)['pred'].sum()
    hl_test_stat = (((observed - expected) ** 2) / expected).sum()
    p_value = 1 - stats.chi2.cdf(hl_test_stat, g - 2)
    return hl_test_stat, p_value, observed, expected

hl_stat, hl_p_value, observed, expected = hosmer_lemeshow_test(y_test, y_pred_prob)
print(f'Hosmer-Lemeshow test Statistic: {hl_stat}')
print(f'Hosmer-Lemeshow test p-value: {hl_p_value}')

# 
if hl_p_value > 0.05:
    print("H0:")
else:
    print("H1:")

# 
groups = range(1, len(observed) + 1)
plt.figure(figsize=(10, 6))
plt.plot(groups, observed, marker='o', linestyle='-', color='b', label='Observed')
plt.plot(groups, expected, marker='o', linestyle='--', color='r', label='Expected')
plt.xlabel('Group')
plt.ylabel('Count')
plt.title('Hosmer-Lemeshow test: Observed vs Expected')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns

# 
np.random.seed(0)
data = pd.DataFrame({
    'value': np.random.randn(60) + np.repeat([1, 2, 3], 20),
    'group': np.repeat(['A', 'B', 'C'], 20)
})

# ANOVA
model = ols('value ~ C(group)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# 
p_value = anova_table["PR(>F)"].iloc[0]
if p_value < 0.05:
    print("H1:")
else:
    print("H0:")
    
# 2x1
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 
sns.boxplot(x='group', y='value', data=data, ax=axes[0])
axes[0].set_title('Box Plot of Different Groups')

# 
group_means = data.groupby('group')['value'].mean().reset_index()
sns.lineplot(x='group', y='value', data=group_means, marker='o', ax=axes[1])
axes[1].set_title('Line Plot of Group Means')

plt.tight_layout()
plt.show()
```

```python
# Define a wider range of degrees of freedom for each distribution
t_dfs = [2, 5, 10, 30, 50]
chi2_dfs = [1, 2, 4, 6, 11]
f_dfs = [(4, 4), (10, 4), (10, 10), (4, 10), (10, 15)]

# Colors for different degrees of freedom
colors = plt.cm.viridis(np.linspace(0, 1, 5))

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Plot t-distribution with different dfs
for df, color in zip(t_dfs, colors):
    dist = stats.t(df=df)
    x = np.linspace(dist.ppf(0.01), dist.ppf(0.99), 100)
    y = dist.pdf(x)
    axs[0, 0].plot(x, y, color=color, lw=2, label=f't-distribution (df={df})')
    axs[0, 0].fill_between(x, y, color=color, alpha=0.2)
axs[0, 0].set_title('t-distribution')
axs[0, 0].legend()

# Plot z-distribution as a reference with different sample sizes
z_dists = [stats.norm(scale=np.sqrt(1/n)) for n in [5, 10, 20, 30, 50]]
for z_dist, color, n in zip(z_dists, colors, [5, 10, 20, 30, 50]):
    x = np.linspace(z_dist.ppf(0.01), z_dist.ppf(0.99), 100)
    y = z_dist.pdf(x)
    axs[0, 1].plot(x, y, color=color, lw=2, label=f'z-distribution (n={n})')
    axs[0, 1].fill_between(x, y, color=color, alpha=0.2)
axs[0, 1].set_title('z-distribution')
axs[0, 1].legend()

# Plot Chi-square distribution with different dfs and expand y-limits
for df, color in zip(chi2_dfs, colors):
    dist = stats.chi2(df=df)
    x = np.linspace(dist.ppf(0.01), dist.ppf(0.99), 100)
    y = dist.pdf(x)
    axs[1, 0].plot(x, y, color=color, lw=2, label=f'Chi-square (df={df})')
    axs[1, 0].fill_between(x, y, color=color, alpha=0.2)
axs[1, 0].set_ylim(0, 0.6)
axs[1, 0].set_title('Chi-square distribution')
axs[1, 0].legend()

# Plot F-distribution with different dfs and expand y-limits
for (dfn, dfd), color in zip(f_dfs, colors):
    dist = stats.f(dfn=dfn, dfd=dfd)
    x = np.linspace(dist.ppf(0.01), dist.ppf(0.99), 100)
    y = dist.pdf(x)
    axs[1, 1].plot(x, y, color=color, lw=2, label=f'F-distribution (dfn={dfn}, dfd={dfd})')
    axs[1, 1].fill_between(x, y, color=color, alpha=0.2)
axs[1, 1].set_ylim(0, 0.6)
axs[1, 1].set_title('F-distribution')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
