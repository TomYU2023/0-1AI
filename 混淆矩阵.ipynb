# 混淆矩阵
混淆矩阵是一个二维矩阵，行表示真实值，列表示预测值。

- TP：True Positive，预测为正例，实际正例
- TN：True Negative，预测为负例，实际负例
- FP：False Positive，预测为正例，实际负例
- FN：False Negative，预测为负例，实际正例

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np

# 设置字体路径
font_path = r'F:\AI_Learning\AI(python)\字体\NotoSansCJK-Bold-6.otf'
font_prop = FontProperties(fname=font_path)

# 创建一个混淆矩阵，标签代替数值
confusion_matrix = np.array([['TP\n真阳性', 'FP\n假阳性'], ['FN\n假阴性', 'TN\n真阴性']])
labels = ['Positive', 'Negative']

# 绘制空白的混淆矩阵
fig, ax = plt.subplots(figsize=(8, 6))
ax.imshow([[0, 0], [0, 0]], cmap='Blues')

# 在矩阵中添加文本
for (i, j), label in np.ndenumerate(confusion_matrix):
    ax.text(j, i, label, ha='center', va='center', fontsize=20, color='black', fontproperties=font_prop)

# 设置轴标签
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels, fontproperties=font_prop)
ax.set_yticklabels(labels, fontproperties=font_prop)

# 添加轴标签
plt.xlabel('Predicted', fontproperties=font_prop)
plt.ylabel('Actual', fontproperties=font_prop)
plt.title('Confusion Matrix', fontproperties=font_prop)

plt.show()


print("TP (True Positive): 真实为正类，预测为正类的样本数")
print("FP (False Positive): 真实为负类，预测为正类的样本数")
print("FN (False Negative): 真实为正类，预测为负类的样本数")
print("TN (True Negative): 真实为负类，预测为负类的样本数")
```

# 评价指标
- Precision：预测为正例的样本中，实际正例的比例
- Recall：实际正例中，预测为正例的比例
- F1-score：Precision和Recall的调和平均值
- Accuracy：预测正确的样本比例
- Confusion Matrix：混淆矩阵，用来评估分类器的性能
- ROC Curve：Receiver Operating Characteristic Curve，ROC曲线，用来评估二分类器的性能
- AUC：Area Under the Curve，ROC曲线下的面积，用来评估二类器的性能


1. **Precision (精确率)**：
   预测为正例的样本中，实际正例的比例。
   $$
   \text{Precision} = \frac{TP}{TP + FP}
   $$

   - **直觉理解**：假设你是一名医生，正在筛查一种罕见的疾病。你预测了10个人有这种疾病，其中8个人确实有这种疾病。精确率就是你预测为阳性（有疾病）的人中，实际有疾病的比例。
   - **类比**：精确率就像是“你喊狼来了，有多少次是真的来了狼”。


```python
# Precision: 预测为正例的样本中，实际正例的比例

TP = 80  # 预测为正例且实际为正例
FP = 20  # 预测为正例但实际为负例

precision = TP / (TP + FP)
print(f"Precision: {precision:.2f}")
```



2. **Recall (召回率)**：
   实际正例中，预测为正例的比例。
   $$
   \text{Recall} = \frac{TP}{TP + FN}
   $$

   - **直觉理解**：在同样的筛查中，你预测了10个人有这种疾病，而实际上有20个人有这种疾病。召回率是实际有疾病的人中，被你正确预测为有疾病的比例。
   - **类比**：召回率就像是“田里有多少狼，你抓到了多少只”。


```python
# Recall: 实际正例中，预测为正例的比例

TP = 80  # 预测为正例且实际为正例
FN = 20  # 预测为负例但实际为正例

recall = TP / (TP + FN)
print(f"Recall: {recall:.2f}")
```


3. **F1-score**：
   Precision和Recall的调和平均值。
   $$
   \text{F1-score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
   $$

   - **直觉理解**：F1-score是精确率和召回率的调和平均。它考虑了精确率和召回率的平衡，是一个综合指标。当你希望你的模型在精确率和召回率之间有一个好的平衡时，F1-score 是一个很好的指标。
   - **类比**：F1-score就像是“在抓狼的过程中，既要确保抓到的都是狼（精确率），又要尽可能多地抓到所有的狼（召回率）”。



4. **Accuracy (准确率)**：
   预测正确的样本比例。
   $$
   \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
   $$

   - **直觉理解**：准确率是所有样本中，预测正确的样本比例。如果你筛查了100个人，预测正确的有90个，那么你的准确率就是90%。
   - **类比**：准确率就像是“所有人中，你预测正确的比例”。



5. **Confusion Matrix (混淆矩阵)**：
   用来评估分类器的性能。混淆矩阵的结构如下：
   $$
   \begin{array}{c|c|c}
       & \text{Predicted Positive} & \text{Predicted Negative} \\
   \hline
   \text{Actual Positive} & TP & FN \\
   \text{Actual Negative} & FP & TN \\
   \end{array}
   $$

   - **直觉理解**：混淆矩阵是一种详细的方式来展示预测结果。它分成四个部分：真正例（TP）、假正例（FP）、假负例（FN）、真负例（TN）。可以帮助你看到具体的预测错误情况。
   - **类比**：混淆矩阵就像是“一个表格，展示了你喊狼来了、狼真的来了、你没喊狼、但狼来了等各种情况”。



6. **ROC Curve (ROC曲线)**：
   Receiver Operating Characteristic Curve，用来评估二分类器的性能。ROC曲线是通过绘制不同阈值下的真阳性率（Recall）与假阳性率（False Positive Rate, FPR）来表示的。
   $$
   \text{FPR} = \frac{FP}{FP + TN}
   $$

   - **直觉理解**：ROC曲线是通过绘制不同阈值下的真阳性率（Recall）与假阳性率（FPR）来表示的。曲线下面积越大，表示模型的性能越好。
   - **类比**：ROC曲线就像是“不同情况下，你在喊狼和没喊狼的各种比例之间的关系”。



7. **AUC (曲线下面积)**：
   ROC曲线下的面积，用来评估二分类器的性能。AUC值在0.5到1之间，值越大表示模型性能越好。

   - **直觉理解**：AUC是ROC曲线下面积。AUC值在0.5到1之间，值越大表示模型性能越好。
   - **类比**：AUC就像是“你在各种情况下喊狼的整体表现分数，越接近1表示表现越好”。

假设我们有一个分类器，用来检测邮件是否为垃圾邮件。以下是混淆矩阵的具体例子：

- $TP$（True Positive）：50封实际是垃圾邮件的邮件被正确分类为垃圾邮件。
- $TN$（True Negative）：100封实际非垃圾邮件的邮件被正确分类为非垃圾邮件。

- $FP$（False Positive）：10封实际非垃圾邮件的邮件被错误分类为垃圾邮件。
- $FN$（False Negative）：5封实际是垃圾邮件的邮件被错误分类为非垃圾邮件。


我们来计算各种指标：
1. **Confusion Matrix (混淆矩阵)**：
   用于评估分类器的性能，具体例子如下：
   $$
   \begin{array}{c|c|c}
       & \text{Predicted Positive} & \text{Predicted Negative} \\
   \hline
   \text{Actual Positive} & TP = 50 & FN = 5 \\
   \text{Actual Negative} & FP = 10 & TN = 100 \\
   \end{array}
   $$
   直观理解：这个矩阵显示了分类器在不同类别上的表现情况。

2. **Precision (精确率/误检率)**：
   精确率是预测为垃圾邮件的邮件中，实际是垃圾邮件的比例。
   $$
   \text{Precision} = \frac{TP}{TP + FP} = \frac{50}{50 + 10} = \frac{50}{60} \approx 0.833
   $$
   直观理解：在所有被分类为垃圾邮件的邮件中，有83.3%实际上是垃圾邮件。

3. **Recall (召回率/漏检率)**：
   召回率是实际是垃圾邮件的邮件中，被正确预测为垃圾邮件的比例。
   $$
   \text{Recall} = \frac{TP}{TP + FN} = \frac{50}{50 + 5} = \frac{50}{55} \approx 0.909
   $$
   直观理解：在所有实际是垃圾邮件的邮件中，有90.9%被正确识别为垃圾邮件。

4. **F1-score**：
   F1-score是精确率和召回率的调和平均值。
   $$
   \text{F1-score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} = 2 \times \frac{0.833 \times 0.909}{0.833 + 0.909} \approx 0.87
   $$
   直观理解：F1-score综合考虑了精确率和召回率，是两者的平衡指标，取值越高表示模型性能越好。

5. **Accuracy (准确率/正确率)**：
   准确率是所有预测正确的邮件占总邮件数的比例。
   $$
   \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} = \frac{50 + 100}{50 + 100 + 10 + 5} = \frac{150}{165} \approx 0.91
   $$
   直观理解：在所有邮件中，有91%的邮件被正确分类。

6. **ROC Curve (ROC曲线)**：
   ROC曲线展示了不同阈值下模型的真阳性率（召回率）与假阳性率（FPR）的关系。
   假阳性率的计算公式：
   $$
   \text{FPR} = \frac{FP}{FP + TN} = \frac{10}{10 + 100} = \frac{10}{110} \approx 0.091
   $$
   直观理解：ROC曲线可以帮助我们选择最合适的分类阈值，以便在不同应用场景中平衡召回率和假阳性率。

7. **AUC (曲线下面积)**：
   ROC曲线下的面积，用来评估二分类器的整体性能，AUC值越接近1，模型性能越好。

通过这个例子，我们可以更直观地理解各个指标的意义及其计算方法。如果有进一步的问题或需要更详细的解释，请告诉我。

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, roc_curve, auc
import matplotlib.pyplot as plt

# 真实标签和预测标签
y_true = np.array([1]*50 + [0]*10 + [1]*5 + [0]*100)
y_pred = np.array([1]*50 + [1]*10 + [0]*5 + [0]*100)

# 计算各个指标
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
accuracy = accuracy_score(y_true, y_pred)
fpr, tpr, _ = roc_curve(y_true, y_pred)
roc_auc = auc(fpr, tpr)

# 打印各个指标
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-score: {f1:.2f}')
print(f'Accuracy: {accuracy:.2f}')
print(f'AUC: {roc_auc:.2f}')

# 绘制混淆矩阵
confusion_matrix = np.array([[50, 10], [5, 100]])
labels = ['Positive', 'Negative']

fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(confusion_matrix, cmap='Blues')

for (i, j), val in np.ndenumerate(confusion_matrix):
    ax.text(j, i, f'{val}\n{["TP", "FP", "FN", "TN"][i*2+j]}', ha='center', va='center', color='black', fontsize=12)

# 设置刻度和标签
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.colorbar(cax)
plt.show()

# 绘制ROC曲线
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

### ROC(用于分类任务)
ROC 的全称是 Receiver Operating Characteristic curve（受试者工作特征曲线）。它是一种图形化表示工具，用于评估分类模型的性能，特别是在二分类问题中。

ROC 曲线绘制了分类模型的 True Positive Rate（真正率，TPR） 对 False Positive Rate（假正率，FPR） 的关系图。TPR 也称为 灵敏度（sensitivity），而 FPR 也称为 1-特异性（1-specificity）。

在 ROC 曲线上：

Y轴 代表 True Positive Rate (TPR)，即敏感度。
X轴 代表 False Positive Rate (FPR)。
理想情况下，分类模型的 ROC 曲线会尽可能接近左上角，即高 TPR 和低 FPR。模型性能的好坏可以通过 ROC 曲线下的面积（AUC，Area Under the Curve）来衡量。AUC 的值介于 0 到 1 之间，值越接近 1 表示模型性能越好。
