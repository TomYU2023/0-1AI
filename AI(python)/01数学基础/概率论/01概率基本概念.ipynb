# 基本概念
## 1. 随机的概念

1. 真随机（True Random）：由自然现象或物理过程产生的随机性，无法通过算法预测。
2. 伪随机（Pseudo-Random）：由算法生成的随机数，看起来像随机数，但实际上是确定性的。

**对比总结**

| **特性**    | **真随机**                         | **伪随机**                             |
|-------------|------------------------------------|----------------------------------------|
| **生成方式** | 自然现象或物理过程                 | 数学算法                               |
| **可预测性** | 不可预测                           | 可预测（如果知道种子和算法）           |
| **重复性**   | 不可重复                           | 可重复（给定相同的种子）               |
| **应用**     | 安全性要求高的领域，如加密学       | 大多数日常计算，如模拟和游戏           |
| **例子**     | 每次掷硬币的结果是完全随机的，没有规律可循       | 调用`random.randint(0, 10)`生成的数看起来是随机的，如果知道算法和种子值可以预测并重复这些随机数           |

## 2. 频率与概率

1. 频率（Frequency）：在一定次数的试验中，某个事件发生次数的比率，没有取值范围。
2. 概率（Probability）：某个事件在理论上发生的可能性，取值范围介于0和1之间。

**对比总结**

| **特性**    | **频率**                                      | **概率**                                      |
|-------------|-----------------------------------------------|-----------------------------------------------|
| **本质**    | 实验结果的比率                                | 理论上的可能性                                |
| **依赖**    | 依赖于实际试验次数                            | 不依赖于试验次数                              |
| **变化性**  | 试验次数少时波动较大，试验次数多时趋于稳定    | 固定值，不随试验次数变化                      |
| **应用**    | 通过重复试验估计概率，统计数据分析            | 用于理论分析和预测，制定概率模型              |
| **符号**     | $ f $ 或 $ \hat{p} $ | $ P $ 或 $ p $    |
| **公式**     | $ f = \frac{\text{事件发生次数}}{\text{总试验次数}} $     | $ P(A) = \frac{\text{事件 A 的可能方式数}}{\text{所有可能方式数}} $ |
| **例子**     | $ f = \frac{60}{100} = 0.6 $（正面出现60次，掷硬币100次）     |  $ P(3) = \frac{1}{6} \approx 0.167 $（掷六面骰子点数为3的概率） |

**频率与概率的关系**
- 在大数定律的框架下，随着试验次数的增加，频率会趋近于理论概率。例如，掷硬币的频率随着试验次数的增加会越来越接近0.5，这表明频率是概率的一种经验估计。

## 3. 概率函数

概率函数的定义：

$$
P(A) = \frac{n(A)}{n(S)}
$$

其中，$P(A)$ 表示事件 $A$ 的概率，$n(A)$ 表示事件 $A$ 的发生次数，$n(S)$ 表示样本空间的总次数。

1. 集合（Set）：用来描述一群具有某种共同特性的对象，集合中的对象称为元素。
2. 样本空间（Sample Space）：所有可能结果的集合。
3. 事件（Event）：样本空间的一个子集



**对比总结**

| 概念       | 定义                                               | 表达式                                               | 例子                                                        | 常用符号和分类                                                   |
|------------|----------------------------------------------------|------------------------------------------------------|-------------------------------------------------------------|------------------------------------------------------------------|
| 集合       | 描述一组具有共同特性的对象。                        | $A = \{a_1, a_2, \ldots, a_n\}$                      | $A = \{1, 2, 3, 4, 5\}$                                     | 1.$\in$ (属于)</br>2.$\notin$ (不属于)</br>3.$\subseteq$ (子集)</br>4.$\cup$ (并集)</br>5.$\cap$ (交集)</br>6.$\setminus$ (差集)|
| 样本空间   | 表示所有可能结果的集合，通常用$S$或$\Omega$表示。    | $\Omega = \{\omega_1, \omega_2, \ldots, \omega_n\}$  | 掷一枚硬币的样本空间：$\Omega = \{H, T\}$                    | 通常用$\Omega$或$S$表示</br>1.离散样本空间</br>2.连续样本空间             |
| 事件       | 样本空间的一个子集，表示感兴趣的一个或多个结果。     | $A \subseteq \Omega$                                 | 事件A表示至少有一个正面：$A = \{HH, HT, TH\}$                | 1.独立事件 ($P(A \cap B) = P(A)P(B)$)</br>2.互斥事件 ($A \cap B = \emptyset$)</br>3.联合事件 ($A \cup B$)</br>4.补事件 ($A^c$ 或 $\overline{A}$)</br>5.不可能事件 $\emptyset$ |

## 4. 事件运算

| 运算       | 定义                                                              | 表达式                                    | 例子                                      | 说明                                      |
|------------|-------------------------------------------------------------------|-----------------------------------------|-------------------------------------------|-------------------------------------------|
| 并事件（加法） | 至少有一个事件发生。                                             | $P(A \cup B) = P(A) + P(B) - P(A \cap B)$ | 掷一次骰子，事件$A$表示掷出偶数，事件$B$表示大于3的数。 | 至少一个事件发生的概率。                  |
| 交事件（乘法） | 同时发生的事件。                                                 | $P(A \cap B)$                           | 掷一次骰子，事件$A$表示掷出偶数，事件$B$表示大于3的数。 | 两个事件同时发生的概率。                  |
| 互斥事件   | 两个事件不能同时发生。                                            | $A \cap B = \emptyset$                 | 掷一次硬币，事件$A$表示正面朝上，事件$B$表示反面朝上。 | $A$和$B$不能同时发生。                    |
| 补事件（取反） | 事件$A$不发生的情况。                                             | $P(A^c) = 1 - P(A)$                    | 掷一次骰子，事件$A$表示掷出偶数，补事件$A^c$表示掷出奇数。 | 事件$A$不发生的概率。                     |
| 差事件（减法）     | 事件$A$发生且事件$B$不发生的情况。                                | $A \setminus B = A \cap B^c$           | 掷一次骰子，事件$A$表示掷出偶数，事件$B$表示大于4的数。$A \setminus B$表示掷出偶数且不大于4的数（2或4）。 | $A$发生且$B$不发生的概率。                |
| 对立事件   | 事件$A$的对立事件表示事件$A$不发生的情况。                        | $A = \overline{A}$                     | 事件$A$：掷出偶数（$A = \{2,4,6\}$）。事件$A$的对立事件$A$：掷出奇数（$A = \{1,3,5\}$）。 | 事件$A$的对立情况。                       |
| 不可能事件 | 在样本空间中不存在任何可能结果的事件。                              | $P(\emptyset) = 0$                     | 掷一次骰子，事件$A$表示掷出7。             | 事件$A$不可能发生。                       |
| 条件概率   | 在事件$B$发生的条件下事件$A$的概率。                             | $P(A \mid B) = \frac{P(A \cap B)}{P(B)}$| 抽到红心牌的概率，条件是这张牌是红色的。   | 已知事件$B$发生，计算$A$发生的概率。      |
| 独立事件   | 两个事件的发生相互独立。                                         | $P(A \cap B) = P(A)P(B)$               | 掷两次硬币，第一次和第二次掷出正面的概率相互独立。 | $A$和$B$的发生不影响对方的发生概率。      |

**性质**


| 定律       | 定义                                                       | 数学表达式                                          | 例子                                                         |
|------------|------------------------------------------------------------|-----------------------------------------------------|--------------------------------------------------------------|
| 交换律     | 两个事件的并或交运算可以交换顺序。                           | $A \cup B = B \cup A$<br>$A \cap B = B \cap A$      | 事件 $A$：掷出偶数<br>事件 $B$：掷出的数大于3               |
| 分配律     | 并运算和交运算之间的分配关系。                               | $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$<br>$A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$ | 事件 $A$：掷出偶数<br>事件 $B$：掷出的数大于3<br>事件 $C$：掷出的数小于等于2 |
| 结合律     | 多个事件的并或交运算可以重新组合。                           | $(A \cup B) \cup C = A \cup (B \cup C)$<br>$(A \cap B) \cap C = A \cap (B \cap C)$ | 事件 $A$：掷出偶数<br>事件 $B$：掷出的数大于3<br>事件 $C$：掷出的数是1或2  |
| 德摩根定律 | 补集运算与并、交运算之间的关系。                             | 并的补集：$ (A \cup B)^c = A^c \cap B^c $ = $ \overline{A \cup B} = \overline{A} \cap \overline{B} $ </br>交的补集: $ (A \cap B)^c = A^c \cup B^c $ = $ \overline{A \cap B} = \overline{A} \cup \overline{B} $| 事件 $A$：掷出偶数<br>事件 $B$：掷出的数大于3               |


```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import matplotlib.patches as patches

# 设置全局字体为SimHei，确保可以显示中文字符
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 创建一个包含四个子图的画布
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# 绘制并集
def plot_union(ax):
    venn = venn2(subsets=(1, 1, 1), set_labels=('A', 'B'), set_colors=('grey', 'grey'), alpha=1.0, ax=ax)
    venn.get_label_by_id('10').set_text(' ')
    venn.get_label_by_id('10').set_color('grey')
    venn.get_patch_by_id('10').set_alpha(0.5)
    venn.get_label_by_id('01').set_text(' ')
    venn.get_label_by_id('01').set_color('grey')
    venn.get_patch_by_id('01').set_alpha(0.5)
    venn.get_label_by_id('11').set_text('A ∪ B')
    venn.get_patch_by_id('11').set_color('grey')
    venn.get_patch_by_id('11').set_alpha(0.5)
    venn.get_patch_by_id('10').set_edgecolor('black')
    venn.get_patch_by_id('01').set_edgecolor('black')
    venn.get_patch_by_id('11').set_edgecolor('black')
    ax.set_title('并', fontsize=14)
    ax.axis('off')

# 绘制交集
def plot_intersection(ax):
    venn = venn2(subsets=(1, 1, 1), set_labels=('A', 'B'), set_colors=('white', 'white'), alpha=1.0, ax=ax)
    venn.get_label_by_id('10').set_text(' ')
    venn.get_label_by_id('01').set_text(' ')
    venn.get_label_by_id('11').set_text('A ∩ B')
    venn.get_patch_by_id('11').set_color('grey')
    venn.get_patch_by_id('10').set_edgecolor('black')
    venn.get_patch_by_id('01').set_edgecolor('black')
    venn.get_patch_by_id('11').set_edgecolor('black')
    ax.set_title('交', fontsize=14)
    ax.axis('off')
    
# 绘制差集
def plot_difference(ax):
    rect = patches.Rectangle((0, 0), 1, 1, edgecolor='black', facecolor='white')
    circle1 = patches.Circle((0.5, 0.5), 0.4, edgecolor='black', facecolor='grey', alpha=0.5)
    circle2 = patches.Circle((0.5, 0.5), 0.2, edgecolor='black', facecolor='white')
    ax.add_patch(rect)
    ax.add_patch(circle1)
    ax.add_patch(circle2)
    ax.text(0.5, 0.75, 'A', horizontalalignment='center', verticalalignment='center', fontsize=14)
    ax.text(0.5, 0.5, 'B', horizontalalignment='center', verticalalignment='center', fontsize=14)
    ax.text(0.95, 0.95, 'Ω', horizontalalignment='center', verticalalignment='center', fontsize=14)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_title('差', fontsize=14)
    ax.axis('off')

# 绘制补集
def plot_complement(ax):
    rect = patches.Rectangle((0, 0), 1, 1, edgecolor='black', facecolor='grey', alpha=0.5)
    circle = patches.Circle((0.5, 0.5), 0.3, edgecolor='black', facecolor='white')
    ax.add_patch(rect)
    ax.add_patch(circle)
    ax.text(0.5, 0.5, 'A', horizontalalignment='center', verticalalignment='center', fontsize=14)
    ax.text(0.95, 0.95, 'Ω', horizontalalignment='center', verticalalignment='center', fontsize=14)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_title('补', fontsize=14)
    ax.axis('off')

# 绘制所有图形
plot_union(axs[0, 0])
plot_intersection(axs[0, 1])
plot_difference(axs[1, 0])
plot_complement(axs[1, 1])

# 显示图形
plt.tight_layout()
plt.show()
```

## 5. 事件关系

| 关系        | 定义                                                                                      | 数学表达式                                               | 例子                                                                                      |
|-------------|-------------------------------------------------------------------------------------------|----------------------------------------------------------|-------------------------------------------------------------------------------------------|
| 包含关系    | 如果事件 $ A $ 中的所有元素都属于事件 $ B $，则称事件 $ A $ 包含于事件 $ B $。       | $ A \subseteq B $ 或 $ B \subseteq A $               | 事件 $ A $：掷出偶数（$ A = \{2, 4, 6\} $），事件 $ B $：掷出大于1的数（$ B = \{2, 3, 4, 5, 6\} $）。$ A \subseteq B $。 |
| 相等关系    | 如果事件 $ A $ 和事件 $ B $ 包含完全相同的元素，则称事件 $ A $ 等于事件 $ B $。       | $ A \subseteq B $ 且 $ B \subseteq A $               | 事件 $ A $：掷出偶数（$ A = \{2, 4, 6\} $），事件 $ B $：掷出偶数（$ B = \{2, 4, 6\} $）。$ A = B $。                        |
| 相交关系    | 如果事件 $ A $ 和事件 $ B $ 有至少一个共同元素，则称事件 $ A $ 和事件 $ B $ 是相交的。 | $ A \cap B \neq \emptyset $                            | 事件 $ A $：掷出偶数（$ A = \{2, 4, 6\} $），事件 $ B $：掷出的数为4或5（$ B = \{4, 5\} $）。$ A \cap B = \{4\} \neq \emptyset $。     |
| 互斥关系    | 如果事件 $ A $ 和事件 $ B $ 没有任何共同元素，则称事件 $ A $ 和事件 $ B $ 是互斥的。   | $ A \cap B = \emptyset $                               | 事件 $ A $：掷出偶数（$ A = \{2, 4, 6\} $），事件 $ B $：掷出奇数（$ B = \{1, 3, 5\} $）。$ A \cap B = \emptyset $。            |
| 对立关系    | 事件 $ A $ 的对立事件表示事件 $ A $ 不发生的情况。                                       | $ A = \overline{B} $                                   | 事件 $ A $：掷出偶数（$ A = \{2, 4, 6\} $）。事件 $ A $ 的对立事件 $ \overline{A} $：掷出奇数（$ \overline{A} = \{1, 3, 5\} $）。     |


```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 设置画布
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_xlim(0, 10)
ax.set_ylim(3, 10)
ax.set_aspect('equal')

# 绘制包含关系
ax.text(1, 9, '包含关系', fontsize=14, weight='bold')
circle1 = patches.Circle((1.5, 8), 0.5, edgecolor='black', facecolor='none')
circle2 = patches.Circle((1.5, 8), 0.8, edgecolor='black', facecolor='none')
ax.add_patch(circle1)
ax.add_patch(circle2)
ax.text(2.5, 8, r'$A \subseteq B$', fontsize=14)

# 绘制相等关系
ax.text(4, 9, '相等关系', fontsize=14, weight='bold')
circle3 = patches.Circle((4.5, 8), 0.5, edgecolor='black', facecolor='none')
circle4 = patches.Circle((4.5, 8), 0.5, edgecolor='black', facecolor='none')
ax.add_patch(circle3)
ax.add_patch(circle4)
ax.text(5.2, 8, r'$A = B$', fontsize=14)

# 绘制相交关系
ax.text(7, 9, '相交关系', fontsize=14, weight='bold')
circle5 = patches.Circle((7.5, 8), 0.5, edgecolor='black', facecolor='none')
circle6 = patches.Circle((8, 8), 0.5, edgecolor='black', facecolor='none')
ax.add_patch(circle5)
ax.add_patch(circle6)
ax.text(8.7, 8, r'$A \cap B \neq \emptyset$', fontsize=14)

# 绘制互斥关系
ax.text(1, 6, '互斥关系', fontsize=14, weight='bold')
circle7 = patches.Circle((1.5, 5), 0.5, edgecolor='black', facecolor='none')
circle8 = patches.Circle((2.5, 5), 0.5, edgecolor='black', facecolor='none')
ax.add_patch(circle7)
ax.add_patch(circle8)
ax.text(3.2, 5, r'$A \cap B = \emptyset$', fontsize=14)

# 绘制对立关系
ax.text(5.5, 6, '对立关系', fontsize=14, weight='bold')
circle9 = patches.Circle((6.5, 5), 0.5, edgecolor='black', facecolor='none')
circle10 = patches.Circle((6.5, 5), 0.8, edgecolor='black', linestyle='dashed', facecolor='none')
ax.add_patch(circle9)
ax.add_patch(circle10)
ax.text(7.5, 5, r'$A = \overline{B}$', fontsize=14)

# 隐藏坐标轴
ax.axis('off')

# 输出图像
plt.show()
```

## 6. 排列与组合

假设有两种颜色的球（红色和蓝色），每种颜色各有两个编号（1和2），即红色球有$A_1$，$A_2$，蓝色球有$B_1$，$B_2$。

| 原理       | 定义                                 | 公式                                        | 应用                                      | 区别                                                                 |
|------------|--------------------------------------|---------------------------------------------|-------------------------------------------|----------------------------------------------------------------------|
| 加法原理   | 完成任务的不同互斥方法数的总和       | $N = n_1 + n_2 + \cdots + n_k$              | 从红色或蓝色球中选一个，$2 + 2 = 4$种方式 | 各方法互斥，结果是总和。</br>选择红色或蓝色球，共4种选择方式。       |
| 乘法原理   | 完成任务的独立步骤数的乘积           | $N = n_1 \times n_2 \times \cdots \times n_k$ | 选择一个红色球和一个蓝色球，$2 \times 2 = 4$种方式 | 各步骤独立，结果是乘积。</br>选择一个红色球和一个蓝色球，共4种组合方式。|
| 排列       | 从$n$个中选$r$个并考虑顺序的方式数  | $P(n, r) = \frac{n!}{(n-r)!}$               | 从$A_1$, $A_2$, $B_1$, $B_2$中选2个并排列 | 考虑顺序，顺序不同视为不同。</br>从4个球中选2个并排列，有12种排列方式。  |
| 组合       | 从$n$个中选$r$个但不考虑顺序的方式数 | $C(n, r) = \frac{n!}{r!(n-r)!}$             | 从$A_1$, $A_2$, $B_1$, $B_2$中选2个      | 不考虑顺序，同样元素组合只算一次。</br>从4个球中选2个，不考虑顺序，有6种组合方式。 |



```python
def add_principle(*events):
    """
    加法原理计算器
    输入多个事件的方式数，返回总方式数
    """
    return sum(events)

def mul_principle(*steps):
    """
    乘法原理计算器
    输入多个步骤的方式数，返回总方式数
    """
    result = 1
    for step in steps:
        result *= step
    return result

def permutation(n, r):
    """
    排列计算器
    计算从n个元素中选r个并考虑顺序的排列数
    """
    from math import factorial
    return factorial(n) // factorial(n - r)

def combination(n, r):
    """
    组合计算器
    计算从n个元素中选r个但不考虑顺序的组合数
    """
    from math import factorial
    return factorial(n) // (factorial(r) * factorial(n - r))

# 示例
add_result = add_principle(2, 2)  # 加法原理示例，2种红球选择 + 2种蓝球选择
mul_result = mul_principle(2, 2)  # 乘法原理示例，选择1个红球 × 选择1个蓝球
perm_result = permutation(4, 2)  # 排列示例，从4个球中选2个并排列
comb_result = combination(4, 2)  # 组合示例，从4个球中选2个

# 输出结果
print(f"加法原理结果: {add_result}")  # 输出: 4
print(f"乘法原理结果: {mul_result}")  # 输出: 4
print(f"排列结果: {perm_result}")     # 输出: 12
print(f"组合结果: {comb_result}")     # 输出: 6
```

## 7. 贝叶斯定理


| 名称             | 定义                                     | 公式                                                     | 应用                                      | 区别                               |
|------------------|------------------------------------------|----------------------------------------------------------|-------------------------------------------|------------------------------------|
| 贝叶斯公式       | 计算已知事件的条件概率                  | $P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}$      | 在医学诊断中结合病史和新症状更新诊断概率   | 利用条件概率反向推断               |
| 全概率公式       | 计算事件的总概率                         | $P(A) = \sum_{i=1}^n P(A \mid B_i) \cdot P(B_i)$         | 计算产品总体次品率                         | 事件划分下的概率求和               |
| 条件概率         | 已知事件$B$发生的条件下$A$发生概率       | $P(A \mid B) = \frac{P(A \cap B)}{P(B)}$                 | 在已知是蓝色球的情况下选到$B_1$的概率      | 依赖另一个事件的发生               |
| 容斥原理         | 计算多个集合并集的元素个数               | $ \mid A \cup B \mid = \mid A \mid + \mid B \mid - \mid A \cap B \mid $                     | 计算两个班级学生的总人数                   | 考虑重叠部分的修正                 |

## 8. 事件独立性


| 概念       | 定义                                                              | 公式                                    | 例子                                      | 说明                                      |
|------------|-------------------------------------------------------------------|-----------------------------------------|-------------------------------------------|-------------------------------------------|
| 事件独立性 | 两个事件$A$和$B$是独立的，当且仅当一个事件的发生不影响另一个事件的发生概率 | $P(A \cap B) = P(A) \cdot P(B)$         | 掷硬币：$P(\text{正面} \cap \text{正面}) = 0.5 \cdot 0.5 = 0.25$ | 掷硬币两次的结果互不影响                |
| 多个事件独立性 | 多个事件$A_1, A_2, \ldots, A_n$是相互独立的，当且仅当任意子集的组合也满足独立性 | $P(A_{i_1} \cap A_{i_2} \cap \ldots \cap A_{i_k}) = P(A_{i_1}) \cdot P(A_{i_2}) \cdot \ldots \cdot P(A_{i_k})$ | 掷骰子：$P(\text{第一次6} \cap \text{第二次6}) = \frac{1}{6} \cdot \frac{1}{6} = \frac{1}{36}$ | 掷骰子两次的结果互不影响               |
| 判断独立性 | 判断两个事件是否独立的方法                                         | 比较$P(A \cap B)$与$P(A) \cdot P(B)$    | -                                         | 如果相等，则独立；否则，不独立            |
| 非独立事件 | 两个事件不满足独立性条件                                          | $P(A \cap B) \neq P(A) \cdot P(B)$      | 从不放回的袋子中取球                     | 取出一个球后的结果影响下一次取球的结果  |
