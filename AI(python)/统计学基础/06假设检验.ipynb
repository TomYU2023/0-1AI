### 假设检验
- 原假设/零假设（Null hypothesis） 斜率为0；H0：一般是希望拒绝的假设。
- 备择假设/替换假设（Alternative hypothesis）斜率不为0；H1：原假设的对立假设。
- α错误/弃真错误：H0为真却被我们拒绝了；这样的概率为α。
- β错误/取伪错误：H1为假却被我们接受了；这样的概率为β。
- 降低犯错的概率提高样本量是方法之一。



假设检验是一种统计方法，用于评估样本数据是否支持特定假设。假设检验的步骤可以分为以下几个关键阶段：

### 假设检验的步骤

1. **提出假设**：
   - **原假设（Null Hypothesis，$H_0$）**：通常表示没有效果或没有差异的假设。
   - **备择假设（Alternative Hypothesis，$H_A$）**：表示有效果或有差异的假设。

2. **选择适当的检验方法**：
   - 根据研究问题、数据类型和样本量选择合适的统计检验方法（如T检验、Z检验、卡方检验等）。

3. **确定显著性水平（Significance Level，$\alpha$）**：
   - 通常选择0.05或0.01作为显著性水平，表示在95%或99%的置信水平下检验假设。

4. **收集数据和计算检验统计量**：
   - 根据样本数据计算检验统计量（如T值、Z值、卡方值等）。

5. **计算P值或查找临界值**：
   - **P值**：计算得到的检验统计量对应的P值。
   - **临界值**：根据显著性水平和检验类型查找相应的临界值。

6. **作出决策**：
   - 比较P值与显著性水平$\alpha$：
     - 如果$P \leq \alpha$，拒绝原假设$H_0$。
     - 如果$P > \alpha$，不拒绝原假设$H_0$。
   - 或者比较检验统计量与临界值：
     - 如果检验统计量超过临界值，拒绝原假设$H_0$。
     - 如果检验统计量不超过临界值，不拒绝原假设$H_0$。

7. **结论和解释**：
   - 根据决策结果得出结论，并解释结论的意义和实际应用。

### 假设检验的详细步骤示例

假设我们想要检验一个新药对降低血压是否有效，以下是具体的假设检验步骤：

#### 步骤1：提出假设
- **原假设（$H_0$）**：新药对降低血压没有显著效果，即新药组和对照组的平均血压相同。
- **备择假设（$H_A$）**：新药对降低血压有显著效果，即新药组的平均血压低于对照组。

#### 步骤2：选择适当的检验方法
- 由于我们比较的是两个独立样本的均值，可以选择独立样本T检验。

#### 步骤3：确定显著性水平
- 选择显著性水平$\alpha = 0.05$。

#### 步骤4：收集数据和计算检验统计量
- 假设我们收集到的数据如下：
  - 新药组：$X_1 = [120, 118, 122, 115, 119]$
  - 对照组：$X_2 = [130, 132, 128, 135, 131]$
- 计算新药组和对照组的样本均值、标准差和样本量：
  - $\bar{X}_1 = 118.8$
  - $\bar{X}_2 = 131.2$
  - $s_1 = 2.68$
  - $s_2 = 2.86$
  - $n_1 = n_2 = 5$
- 计算T值：
$$
t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}} = \frac{118.8 - 131.2}{\sqrt{\frac{2.68^2}{5} + \frac{2.86^2}{5}}} \approx -9.55
$$

#### 步骤5：计算P值或查找临界值
- 根据计算的T值和自由度$df = n_1 + n_2 - 2 = 8$，查找T分布表或使用统计软件计算P值。
- 假设P值计算为$P < 0.001$。

#### 步骤6：作出决策
- 比较P值与显著性水平$\alpha$：
  - 由于$P < 0.001 < 0.05$，拒绝原假设$H_0$。

#### 步骤7：结论和解释
- 结论：我们有足够的证据表明新药对降低血压有显著效果。
- 解释：新药组的平均血压显著低于对照组，表明新药在降低血压方面有效。

通过这些步骤，我们可以系统地进行假设检验，评估样本数据是否支持特定假设。
