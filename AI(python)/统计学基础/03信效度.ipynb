### 信度（Reliability）：用来衡量工具的稳定度，即重复施以相同的测量，可以获大致相同的结果，即可重复性。
- 因子载荷（Factor loading）:每个题目的信度
- 组合信度（Composite reliability）：维度的信度
- 内部一致性（Cronbach’s α）：维度中题目回答的一致性

### 效度（Validity）:测量内容与实际情形的接近程度，即准确性。
- 建构效度：
    - 区别效度（Discriminant validity）：维度间的区别
    - 收敛效度（Convergent validity）：题目间的相关


### 信度
在不同的时间，地点，测量同一群体的人，得到跟之前做的测量差不多的答案。（可不可以重复）  
以英语测试为例：  
背景：用一套信效度高的方法测量大家的英语水平，这里的信效度高指的就是某个小孩英文考试成绩高不是巧合，而是可以反复出现的成绩点数（信度），并且确实测量到了他的英语水平（效度）。  
课前务必明确：反复考试（信度），是不是考英语题（效度）  

 1. 信度高，效度低：用数学卷子测大家英文水平（明明英文该 90 分水平的孩子，因为数学考了 20 分，虽然重复测量以后数学可能都是 20 分左右，都是你说他英文差是不合理的。这就是我们说的效度低，没测准）。
2. 信度低，效度低：用抽签的方式测大家英文水平。（不能重复得到稳定分值，也测不准真实水平。） 
3. 信度低，效度高：**几乎不存在**。因为效度高，信度必然高。（你不可能因为明明用英文卷子测大家英文水平，却不能重复测量。该考八九十分的孩子重复考这套卷子依然会是八九十分。）
4. 信度高，效度高：用英文卷子测大家的英文水平。（可以重复测量他们真实的英文水平。）



[[误差]]：导致信效度低的原因就是误差。

---
信度（只受随机误差的影响） 
1. 重测信度：同一量表，同一批群体 (一般设置编号题记住每个人的身份，第二次测量时必须全员到位；一般用于成年人)，不同时间 (2~4 周以内，最长不能超过 6 个月，时间越短，信度越高，反之亦然)，2 次施测，求相关；具体公式百度，或者用软件计算，比如 SPSS；实质上测稳定性系数 coefficient of stability（用皮尔逊相关系数也称为积差相关或积矩相关来表示；一般结果要求相关大于 0.7）；除此以外间隔时间太短的重测还需考虑练习/记忆效应，被试者可能掌握答题技巧或者记住题目答案而导致获得的信度不准确(信度大于 0.7～小于 0.9 是最合适的，如果过高就证明题目高度相似，才会出现重复的答案；比如 10 个题都是 1+1 等于几，一加一等于几，一乘一等于几这样类似的题目，那测量结果显然是可以高度重复的)；比较麻烦一般不推荐。 
2. 复本信度：2 份测量目的一样的问卷，同时测量同一批群体，然后计算皮尔逊相关系数（除非有先行研究参考先行研究的，不然不推荐，非常麻烦。比如找到目的一样的问题，数量也必须一样，但是表达方式还不同，还得满足标准差，平均数等，所以一般不使用该方法）。 
3. 分半信度：将一个维度中的题目分成对等的2份（比如按奇偶数分开）测量回答者在这这2个测量中得分的相关系数。（比较麻烦不推荐） 
4. Cronbach alpha 系数（非常推荐）：用于测量所有题目是不是测量的同一心理特征。（不能直接测2阶维度，要拆解成1阶模型进行单独测量；1阶维度的意思是比如，测英文听力，那么就需要很多具体关于英文听力的考题去测量；2阶维度的意思是比如英文水平，下面包含了听说读写4个1阶维度，各维度下有具体的测量题目）；一般 Cronbach alpha 系数要求大于0.7~0.9。 



### 效度
效度（随机误差和系统误差都影响效度；比如一个小孩虽然英文能力很好，但是由于该题目逻辑思维性太强（系统误差），也有可能不能展示出语言水平，也就是看得懂单词，不理解深层的逻辑，即不能准确的测量出效度）   
1. 内容效度：问卷的题目是否能够测量出想要的内容。比如具有听说读写的英文试卷测量英文水平。（常见的方法是请专家来判断(涉及专家效度；这种方法依赖专家的逻辑所以也叫逻辑效度，不推荐，具体证明方法很麻烦)问卷内容是否全面，或者查找先行文献(推荐)中专家们提及的范围） 
2. 效标（关联）效度：效标是指研究者希望预测的行为。测验分数多大程度上能反应希望测量的维度的表现。比如已知英文听说读写的成绩和英文能力有关，那么你出的英文试卷多大程度上能反应这个人的英文水平。效标效度可以分为 2 种：   
	a. 预测效度：先收集预测分数（比如高考成绩），过一段时间再收集效标分数（比如大学学习成绩），从而用高考分数预测大学的学习能力的有效性如何，即高考成绩是否能测准大学时的学习能力。   
	b. 同时效度：测验实施（比如询问学生是否喜欢上网）和效标测量（观察他们课后表现或者访谈周围同学）同时进行，用于推测现有的行为特征（推测他们是否有效地回答了问卷）；效标测量首先必须可靠，稳定；单独的定量研究一般不使用。   
3. 构念效度：测量结果是否能解释这一构念/维度。比如；英文的听说读写是否能有效地解释英文能力这个构念/维度。   
	a. 聚合效度（不常用）：比如不同研究者开发了不同的英文测试问卷，测验结果都是正相关，就证明有聚合效度。   
	b. 区别效度（常用）：比如测量英文能力的题项之间有所不同则证明每个问题不一样，当然我们希望一套英文试卷的题目是多样化的，不完全一样的题目。  

### 误差
都有哪些种类的误差？
- 结论：2种。抽样误差和非抽样误差。

什么是抽样误差？
- 结论：抽样的随机性引起的样本结果与总体真值之间的差异。

什么是非抽样误差？
- 结论	：除了抽样误差都属于非抽样误差。

非抽样误差具体都由哪些情况引发呢？
- 结论：抽样框误差、回答误差、无回答误差、调查员误差、以及测量误差。

举例：
- 抽样框误差：用了已毕业的学生名单（抽样框），抽样大学在籍学生群体。
- 回答误差：回答者因为各种原因没有真实作答。
- 无回答误差：回答者拒绝作答。
- 调查员误差：调查员粗心、或者个人原因造成的误差。
- 测量误差：测量工具本身有误差，比如想测对方英文能力，结果问卷设置的是数学问题。
