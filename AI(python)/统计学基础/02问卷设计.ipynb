### 问卷为了了解什么？
- 事实：您个人使用的手机品牌
- 知识：抽烟的危害
- 态度：满意程度
- 行为：每天使用电脑的时间

### 定量研究的问卷必须是封闭式问卷
- 是非题
- 选择题
- 多选题

量表设计一般选用李克特量表、语义差异量表

每个维度的题目至少3个以上


### 问卷设计原则：
- 参考他人研究相关题目
- 采用封闭式题型
- 题目文字不能太长
- 不要设计反向题
- 衡量尺度大一些好（5或7度）
- 一个维度不要超过7题
- 题目简短直觉
- 避免生僻词
- 避免一个问题问两件事
- 避免明示或者暗示
- 不要侵犯隐私


### 学术型定义和操作型定义

- 学术型定义：界定研究变量和维度名词的意义，即名词解释（比较抽象的概念）。
- 操作型定义：将抽象的概念转化成具体可测量的过程（比较具体的解释）。
- 形成型指标：形成维度的指标；删除指标可能影响维度含义。
- 反应型指标：通过维度能反应出的指标；删除指标不影响维度含义。



### 抽样方法


内生性问题：
- 共同方法偏误：同样的调查方法
- 选择性偏误：抽样不够随机
- 社会期望偏误：答者潜意识回答调查者希望的答案
