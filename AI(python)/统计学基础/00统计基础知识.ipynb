# 概率与统计
概率是什么？
事件发生的可能性大小。

统计是什么？
数据估计与预测的方法。

那么，统计与概率论有什么关系？

## 统计是概率的逆向操作

### 概率：
杯中有三个红球、两个蓝球，红球的概率是3/5


### 统计：
得到一只蓝球、一只红球，求杯中有多少种可能


假如红球个数X服从λ=5的泊松分布，即：

$$
P(X = k) = \frac{\lambda^k}{k!} e^{-\lambda}
$$
其中，
- **$X$**：随机变量，表示事件数量
- **$k$**：事件数量的具体值（非负整数）
- **$\lambda$**：平均事件数量
- **$k!$**：$k$的阶乘
- **$e$**：自然对数的底数（约2.718）

$$
P(X = k) = \frac{5^k}{k!} e^{-5}
$$


这个公式表示的是在一个平均发生5次事件的时间段内，恰好发生$k$次事件的概率。

然而，现实生活中我们并不知道$\lambda$的值是多少。只能用均值估计$\lambda$。
$$
\hat{\lambda}=\frac{\sum_{i=1}^{n}x_i}{n}
$$


另外一个问题就是，我们为什么能知道它就符合泊松分布而不是别的呢？

实际上我们**必须根据经验或者文献**来假设它是泊松分布，然后通过“统计”来检验我们的假设。
