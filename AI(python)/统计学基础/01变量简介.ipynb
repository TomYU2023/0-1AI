

研究角色分类：  
自变量（解释）:独立于系统中
- 纯自变量
- 干扰（调节）变量
- 控制变量

因变量（被解释）：被系统中其他变量影响
- 纯因变量
- 中介变量

1. **自变量（Independent Variable）**：独立于系统中的变量，如温度、湿度、时间等。
2. **因变量（Dependent Variable）**：被系统中其他变量影响，如空气污染、空气污染指数、空气污染等级等。
3. **中介变量（Mediator）**：解释两变量之间的关系是“如何” 及“为何”发生的过程。
4. **调节/干扰变量（Moderator Variable）**：可以了解在不同情境下自变量对因变量变量的影响有何不同。
5. **控制变量（ControlVariable）**：自变量对因变量的影响，对给定的条件和目标，施加控制的过程和行为。
---

自变量和因变量的同义词：
- 独立变量（Independent）→依变量（Dependent）
- 预测变量（Predictor）→准则变量（Criterion）
- 原因变量（Cause）→效果变量（Effect）
- 前置变量（Antecedent）→结果变量（Consequence）
- 外生变量（Exogenous）→内生变量（Endogenous）
- 预测变量（Predictor）→反应变量（Response）
- 特征向量（Feature）→标签(Label)

其他变量的同义词：
- 中介变量：Mediator, Intervening, Intermediate, Process, Mediating, Surrogate variable
- 调节/干扰变量：Moderator variable
- 控制变量/混肴变量/扰动变量/额外变量/伴随变量：作用：Control variable, Covariate, Confounding, Nonsense, Extraneous,Concominant variable
---
观测方式分类：
- 潜变量（Latent variable）→显变量（Manifest variable）
- 不可观察变量（Unobservable variable）→可观察变量（Observable variable）
- 因素/因子（Factor）→指标（Indicator）
- 维度（Dimensions）→项目（Item）
- 元件（Component）
- 构念（Construct）
- 内部属性（Internal attributes）




- 预测变量(IV)及效果变量(DV)，是否有因果关系? 
- 如果有，这个过程如何产生? 
- 是否考虑了所有的过程(所有的中介效果)? 
- 如何评估这些中介效果的有效性? 
- 哪些中介效果比较具有影响力?
- 中介变量的引进时机在于自变量与因变量之间的关系是比较强烈的情形下。
- 中介效果的关系是乘法表示；比如：中介效果=疫情→缺乏娱乐的路径X缺乏娱乐→抑郁的路径




- 调节变量可以是质性变量（类别）， 如性别、班级。
- 调节变量也可以是量化变量（连续）， 如涉入程度，可能影响自变量与因变量之间的强度及/或方向。
- 调节变量适用于X→Y的关系较弱或不一致的时候




- 理论上与因变量有相关的变量， 但却不是本研究的重点。
- 控制变量不可以使用人口统计变量， 如性别、年龄、教育程度
- 应用特定、经过良好理论探索的方式 来纳入控制变量。不可以只是简单的 说明“以前有研究用过”或“变量和结果有关”。
- 在一个研究中，研究人员希望尽量控制一些外在的影响因素不变，以求得研究的准确性，称为纯化原则。
- 实验设计在实验室中比较可以控制环境变量，称为实验控制。
- 社会科学中，无法控制某些环境因素不变，因此,只能排除它的影响， 这种方法称为统计控制。

### 名目尺度（Nominal scale）：性别、职业
二项计数、多项计数数据，无法比较大小

### 顺序尺度（Ordinal scale）：教育程度、收入
有一定顺序，可以分出程度高低

因此，类别尺度切割没有意义，于是不会有均值、方差、标准误、偏态、峰度等

### 区间尺度（Interval scale）：李克特量表，语义量表
数值具有相同距离没有零点，即可以有负值

### 比例尺度（Ratio scale）：价格、年龄
数值具有相同距离且有零点，即无负值

因此，连续尺度切割有意义，于是会有均值、方差、标准误、偏态、峰度等

### 常见词
- Nominal：3个分类以上的名目尺度
- Binary/Dichotomous：2个分类以上的名目尺度、顺序尺度
- Category/Ordinal：3个分类以上的有序尺度的非正太分布
- Normal Scale：5个分类以上的有序尺度、区间尺度的正态分布
