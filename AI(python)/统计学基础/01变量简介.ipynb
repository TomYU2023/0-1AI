
### 依据研究角色分类
- 自变量（名目尺度、顺序尺度、区间尺度、比例尺度）
- 因变量（名目尺度、顺序尺度、区间尺度、比例尺度）
- 中介变量（名目尺度、顺序尺度、区间尺度、比例尺度）
- 调节变量（名目尺度、顺序尺度、区间尺度、比例尺度）
- 控制变量（名目尺度、顺序尺度、区间尺度、比例尺度）

---



研究角色分类：  
自变量（解释）:独立于系统中
- 纯自变量
- 干扰（调节）变量
- 控制变量

因变量（被解释）：被系统中其他变量影响
- 纯因变量
- 中介变量

1. **自变量（Independent Variable）**：独立于系统中的变量，如温度、湿度、时间等。
2. **因变量（Dependent Variable）**：被系统中其他变量影响，如空气污染、空气污染指数、空气污染等级等。
3. **中介变量（Mediator）**：解释两变量之间的关系是“如何” 及“为何”发生的过程。
4. **调节/干扰变量（Moderator Variable）**：可以了解在不同情境下自变量对因变量变量的影响有何不同。
5. **控制变量（ControlVariable）**：自变量对因变量的影响，对给定的条件和目标，施加控制的过程和行为。
---

自变量和因变量的同义词：
- 独立变量（Independent）→依变量（Dependent）
- 预测变量（Predictor）→准则变量（Criterion）
- 原因变量（Cause）→效果变量（Effect）
- 前置变量（Antecedent）→结果变量（Consequence）
- 外生变量（Exogenous）→内生变量（Endogenous）
- 预测变量（Predictor）→反应变量（Response）
- 特征向量（Feature）→标签(Label)

其他变量的同义词：
- 中介变量：Mediator, Intervening, Intermediate, Process, Mediating, Surrogate variable
- 调节/干扰变量：Moderator variable
- 控制变量/混肴变量/扰动变量/额外变量/伴随变量：作用：Control variable, Covariate, Confounding, Nonsense, Extraneous,Concominant variable
---
观测方式分类：
- 潜变量（Latent variable）→显变量（Manifest variable）
- 不可观察变量（Unobservable variable）→可观察变量（Observable variable）
- 因素/因子（Factor）→指标（Indicator）
- 维度（Dimensions）→项目（Item）
- 元件（Component）
- 构念（Construct）
- 内部属性（Internal attributes）




- 预测变量(IV)及效果变量(DV)，是否有因果关系? 
- 如果有，这个过程如何产生? 
- 是否考虑了所有的过程(所有的中介效果)? 
- 如何评估这些中介效果的有效性? 
- 哪些中介效果比较具有影响力?
- 中介变量的引进时机在于自变量与因变量之间的关系是比较强烈的情形下。
- 中介效果的关系是乘法表示；比如：中介效果=疫情→缺乏娱乐的路径X缺乏娱乐→抑郁的路径




- 调节变量可以是质性变量（类别）， 如性别、班级。
- 调节变量也可以是量化变量（连续）， 如涉入程度，可能影响自变量与因变量之间的强度及/或方向。
- 调节变量适用于X→Y的关系较弱或不一致的时候




- 理论上与因变量有相关的变量， 但却不是本研究的重点。
- 控制变量不可以使用人口统计变量， 如性别、年龄、教育程度
- 应用特定、经过良好理论探索的方式 来纳入控制变量。不可以只是简单的 说明“以前有研究用过”或“变量和结果有关”。
- 在一个研究中，研究人员希望尽量控制一些外在的影响因素不变，以求得研究的准确性，称为纯化原则。
- 实验设计在实验室中比较可以控制环境变量，称为实验控制。
- 社会科学中，无法控制某些环境因素不变，因此,只能排除它的影响， 这种方法称为统计控制。

#### 类别变量
- **名目尺度（Nominal scale）**：数据是分类的，没有内在的顺序。
  - **例子**：性别、职业
  - **特点**：二项计数、多项计数数据，无法比较大小。
  - **区别**：性别不能比大小，假设女为0，男为1。性别的**名目**无法比较大小。

- **顺序尺度（Ordinal scale）**：数据是有顺序的，可以排序。
  - **例子**：教育程度、收入
  - **特点**：有一定顺序，可以分出程度高低。
  - **区别**：教育程度可以比大小，假设高中为1，本科为2，博士为3。教育程度的**顺序**可以比大小。

#### 连续变量
- **区间尺度（Interval scale）**：数据有顺序，且间隔相等，但零点是人为设定的。
  - **例子**：年份、李克特量表
  - **特点**：数值具有相同距离，人为假设一个零点。负数有意义。
  - **区别**：年份是人为设定的公元元年0点，人类不知道宇宙起源的0点。无绝对0点，因此只能关注年份的**区间**关系。比如200年比100年早100年，但是100年不是200年的两倍，因为你不知道起源，本质上100年200年的叫法其实相对宇宙时间都是错误的，是人为规定的。

- **比例尺度（Ratio scale）**：数据有顺序，间隔相等，且有绝对零点。
  - **例子**：价格、年龄
  - **特点**：数值具有相同距离，自然产生一个零点，可以进行乘法和除法运算。负数无意义。
  - **区别**：年龄是自然产生的0点，但是人类知道0点是出生。有绝对0点，因此可以计算年龄的**比例**关系。比如20岁是10岁的2倍。




**常见词**
- Nominal：3个分类以上的名目尺度
- Binary/Dichotomous：2个分类以上的名目尺度、顺序尺度
- Category/Ordinal：3个分类以上的有序尺度的非正太分布
- Normal Scale：5个分类以上的有序尺度、区间尺度的正态分布

## 不同尺度的检验方法

![#ffcccc](https://via.placeholder.com/15/ffcccc/000000?text=+) **这里只是大致举例，根据不同的情况不同可能还有其他方法选择。**

### 卡方检验（Chi-square Test）
- **直觉理解**: 检查两个（或多个）类别变量之间是否存在显著关联。
- **计算公式**: 
  $$ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} $$
  - $O_i$: 实际观察到的频数
  - $E_i$: 期望频数

**例子**: 不同性别的学生（自变量-类别变量-名目尺度）对喜欢数学课是否（因变量-类别变量-名目尺度）有差异？

### 单因素方差分析（ANOVA）
- **直觉理解**: 比较三个或更多组的均值是否有显著差异。
- **计算公式**: 
  $$ F = \frac{MS_{between}}{MS_{within}} $$
  - $MS_{between}$: 组间均方
  - $MS_{within}$: 组内均方

**例子**: 不同年级的学生（自变量-类别变量-名目尺度）期末考试的平均成绩（因变量-连续变量-区间尺度）是否有差异？

### 曼-惠特尼U检验（Mann-Whitney U Test）
- **直觉理解**: 比较两组数据的分布是否有显著差异，适用于非正态分布数据。
- **计算公式**: 
  $$ U = n_1 n_2 + \frac{n_1 (n_1 + 1)}{2} - R_1 $$
  - $n_1$: 第一组样本数
  - $n_2$: 第二组样本数
  - $R_1$: 第一组的秩和

**例子**: 不同年级的学生（自变量-类别变量-顺序尺度）学习态度（因变量-类别变量-顺序尺度）有差异吗？

### Kruskal-Wallis检验
- **直觉理解**: 比较三个或更多独立样本的秩和，适用于非正态分布数据。
- **计算公式**: 
  $$ H = \frac{12}{N(N+1)} \sum \frac{R_i^2}{n_i} - 3(N+1) $$
  - $N$: 总样本数
  - $R_i$: 第i组的秩和
  - $n_i$: 第i组的样本数

**例子**: 不同年级的学生（自变量-类别变量-顺序尺度）总考试成绩（因变量-连续变量-区间尺度）是否有差异？

### 逻辑回归分析（Logistic Regression）
- **直觉理解**: 预测二分类因变量，解释一个或多个自变量对因变量的影响。
- **计算公式**: 
  $$ \log \left( \frac{p}{1-p} \right) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_k x_k $$
  - $p$: 事件发生的概率
  - $\beta_i$: 回归系数
  - $x_i$: 自变量

**例子**: 学生的期末考试成绩（自变量-连续变量-区间尺度）是否通过考试（因变量-类别变量-名目尺度）？

### 有序逻辑回归（Ordinal Logistic Regression）
- **直觉理解**: 预测有序多分类因变量，解释一个或多个自变量对因变量的影响。
- **计算公式**: 
  $$ \log \left( \frac{P(Y \leq j)}{P(Y > j)} \right) = \alpha_j - (\beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_k x_k) $$
  - $P(Y \leq j)$: 因变量不超过第j类别的概率
  - $\alpha_j$: 截距
  - $\beta_i$: 回归系数
  - $x_i$: 自变量

**例子**: 学生的期末考试成绩（自变量-连续变量-区间尺度）与他们的成绩等级（因变量-类别变量-顺序尺度）有关吗？

### 皮尔逊相关系数（Pearson Correlation）
- **直觉理解**: 衡量两个连续变量之间的线性关系。
- **计算公式**: 
  $$ r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}} $$
  - $x_i$: 第i个自变量值
  - $y_i$: 第i个因变量值
  - $\bar{x}$: 自变量的均值
  - $\bar{y}$: 因变量的均值

**例子**: 学生的学习时间（自变量-连续变量-区间尺度）与他们的考试成绩（因变量-连续变量-区间尺度）有关系吗？
